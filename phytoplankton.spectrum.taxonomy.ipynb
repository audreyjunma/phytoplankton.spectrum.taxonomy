## Relevant Information

### Background Information

Diminishing sea ice alters the spectral quality of light penetrating polar oceans, thereby impacting photosynthetic organisms and the broader marine ecosystem. The article highlights that as sea ice thins, the under-ice light field collapses from a broad spectrum to a much narrower, blue dominated one. The authors used radiative transfer modelling to reproduce ice covered versus open water spectra and argued that this spectral compression will reshuffle the balance of photosynthetic organisms and thus the biological carbon pump.
The blue light favors smaller phytoplankton in open water conditions, but not much was discussed in terms of ice-covered conditions. However, the SIPEX-2 study did show that at snow depths exceeding 20cm blue-green light was disproportionately suppressed, confirming the Soja-Wozniak study that under ice light is broad spectrum while thinning ice collapses the spectrum. This shift in light quality is critical as it can influence the physiological responses of phytoplankton, such as their growth rates, community composition, and more. 
This spectral compression appears to have a strong effect on phytoplankton communities and export efficiency, but the consequences remain largely unexplored in the Southern Ocean. By adding optics to our current study we could provide further insight into how climate induced changes in sea ice affect carbon sequestration processes in the Southern Ocean.

### Loading + Compiling Data

```python
# create function to load data: under-ice BGC-Argo (chla + bbp) using sprof index + per-day OISST

# weddell sea box 
LON_MIN, LON_MAX = -60.0, 20.0
LAT_MIN, LAT_MAX = -78.0, -62.0
YEARS = [2020, 2021]
ICE_THRESHOLD = 0.15
OUTDIR = "./argo_underice_outputs"; os.makedirs(OUTDIR, exist_ok=True)

# sprof index
GDAC_URLS = [
    "https://data-argo.ifremer.fr/argo_synthetic-profile_index.txt.gz",
    "https://usgodae.org/ftp/outgoing/argo/argo_synthetic-profile_index.txt.gz",
]

def download_sprof_index():
    last = None
    for url in GDAC_URLS:
        try:
            print(f"Downloading Sprof index: {url}")
            r = requests.get(url, timeout=120); r.raise_for_status()
            raw = gzip.decompress(r.content)
            names = ["file","date","latitude","longitude","ocean","profiler_type",
                     "institution","parameters","parameter_data_mode","date_update"]
            df = pd.read_csv(io.BytesIO(raw), comment="#", header=None, sep=",",
                             names=names, dtype=str, engine="python")
            print(f"Index rows: {len(df)}")
            return df
        except Exception as e:
            print(" failed:", e); last = e
    raise last

def filter_region_time_chlabpp(df, year):
    df["latitude"]  = pd.to_numeric(df["latitude"], errors="coerce")
    df["longitude"] = pd.to_numeric(df["longitude"], errors="coerce")
    df["longitude"] = ((df["longitude"] + 180.0) % 360.0) - 180.0
    df["date"] = pd.to_datetime(df["date"], format="%Y%m%d%H%M%S", utc=True, errors="coerce")

    t0 = pd.Timestamp(f"{year}-10-01", tz="UTC")
    t1 = pd.Timestamp(f"{year}-10-31 23:59:59", tz="UTC")
    m_time = df["date"].between(t0, t1, inclusive="both")
    m_reg  = (df["latitude"].between(LAT_MIN, LAT_MAX, inclusive="both") &
              df["longitude"].between(LON_MIN, LON_MAX, inclusive="both"))

    params = df["parameters"].fillna("").str.upper()
    has_chl = params.str.contains(r"\bCHL[A]?\b|CHLORO|CHLOROPHYLL", regex=True)
    has_bbp = params.str.contains(r"\bBBP\b|BBP\d{3}|BETA_BACKSCATTER", regex=True)

    dfg = df[m_time & m_reg & has_chl & has_bbp].copy()

    def parse_wmo(p): 
        try:
            parts = p.split("/")
            return parts[3] if len(parts)>=4 and parts[1]=="dac" else ""
        except: return ""
    def parse_cycle(p):
        m = re.search(r"_([0-9]{3,4})\.nc$", p)
        return int(m.group(1)) if m else np.nan

    dfg["wmo_id"] = dfg["file"].astype(str).apply(parse_wmo)
    dfg["cycle_number"] = dfg["file"].astype(str).apply(parse_cycle)
    dfg = dfg.rename(columns={"date":"datetime","latitude":"lat","longitude":"lon"})
    return dfg

# per-day OISST sampler (ncei fallback if aoml doesn't work)
def sample_oisst_ice_daily(times, lats, lons, pad_deg=0.5):
    """
    query a small OISST 'ice' tile per unique UTC date using 'time=' equality.
    tries AOML first (180..180 grid), then NCEI (0..360 grid). Returns fraction 0..1.
    """
    print("USING per-day OISST sampler")
    times = pd.to_datetime(times, utc=True)
    lats  = np.asarray(lats, dtype="float64")
    lons  = np.asarray(lons, dtype="float64")
    out   = np.full(lats.shape[0], np.nan, dtype="float64")

    # use .strftime directly
    dates = times.strftime("%Y-%m-%d") if isinstance(times, pd.DatetimeIndex) \
            else pd.Series(times).dt.strftime("%Y-%m-%d")

    def fetch_aoml(date_str, lon0, lon1, lat0, lat1):
        e = ERDDAP(server="https://erddap.aoml.noaa.gov/hdb/erddap", protocol="griddap")
        e.dataset_id = "SST_OI_DAILY_1981_PRESENT_HL"
        e.variables = ["ice"]
        for hh in ("00","12"):
            e.constraints = {"time=": f"{date_str}T{hh}:00:00Z",
                             "latitude>=": float(lat0), "latitude<=": float(lat1),
                             "longitude>=": float(lon0), "longitude<=": float(lon1)}
            try:
                e.griddap_initialize(); ds = e.to_xarray()
                if "lat" in ds and "latitude" not in ds: ds = ds.rename({"lat":"latitude"})
                if "lon" in ds and "longitude" not in ds: ds = ds.rename({"lon":"longitude"})
                if ds.sizes.get("time",0)>0: return ds
            except: 
                pass
        return None

    def fetch_ncei(date_str, lon0, lon1, lat0, lat1):
        e = ERDDAP(server="https://www.ncei.noaa.gov/erddap", protocol="griddap")
        e.variables = ["ice"]
        for dsid in ("ncdc_oisst_v2_avhrr_by_time_zlev_lat_lon",
                     "ncdc_oisst_v2_avhrr_amsr_by_time_zlev_lat_lon"):
            for hh in ("00","12"):
                try:
                    e.dataset_id = dsid
                    e.constraints = {"time=": f"{date_str}T{hh}:00:00:00Z" if hh=="00" else f"{date_str}T12:00:00Z",
                                     "latitude>=": float(lat0), "latitude<=": float(lat1),
                                     "longitude>=": float(lon0), "longitude<=": float(lon1)}
                    e.griddap_initialize(); ds = e.to_xarray()
                    if "lat" in ds and "latitude" not in ds: ds = ds.rename({"lat":"latitude"})
                    if "lon" in ds and "longitude" not in ds: ds = ds.rename({"lon":"longitude"})
                    if "zlev" in ds.dims and ds.dims["zlev"]==1: ds = ds.isel(zlev=0)
                    if "depth" in ds.dims and ds.dims["depth"]==1: ds = ds.isel(depth=0)
                    if ds.sizes.get("time",0)>0: return ds
                except:
                    continue
        return None

    # group points by date + fetch a small tile per date
    for date_str, idx in pd.Series(range(len(lats))).groupby(dates):
        ii = np.asarray(list(idx), int)
        lat0, lat1 = float(np.nanmin(lats[ii]) - pad_deg), float(np.nanmax(lats[ii]) + pad_deg)

        # first try: aoml (180..180)
        lons180 = ((lons[ii] + 180.0) % 360.0) - 180.0
        lon0a, lon1a = float(np.nanmin(lons180) - pad_deg), float(np.nanmax(lons180) + pad_deg)
        ds = fetch_aoml(date_str, lon0a, lon1a, lat0, lat1)

        # fallback: ncei (0..360)
        if ds is None:
            lons0360 = np.where(lons[ii] < 0, lons[ii] + 360.0, lons[ii])
            lon0n, lon1n = float(np.nanmin(lons0360) - pad_deg), float(np.nanmax(lons0360) + pad_deg)
            if lon1n < lon0n: lon0n, lon1n = 0.0, 359.875
            ds = fetch_ncei(date_str, lon0n, lon1n, lat0, lat1)

        if ds is None:
            continue

        ice = ds["ice"]
        la  = ice["latitude"].values
        lo  = ice["longitude"].values

        # time index to avoid tz mismatches, if single slice pick index 0, else choose nearest by abs difference
        if ice.sizes.get("time", 0) <= 1:
            ti = 0
        else:
            tvals = pd.to_datetime(ice["time"].values)  
            target = np.datetime64(date_str)            
            ti = int(np.argmin(np.abs(tvals.values.astype("datetime64[ns]") - target.astype("datetime64[ns]"))))

        for i in ii:
            # match longitude wrap to dataset
            if float(lo.min()) < 0:   # AOML grid (180..180)
                xx = ((lons[i] + 180.0) % 360.0) - 180.0
            else:                     # NCEI grid (0..360)
                xx = lons[i] if lons[i] >= 0 else lons[i] + 360.0
            yi = int(np.abs(la - lats[i]).argmin())
            xi = int(np.abs(lo - xx).argmin())
            val = ice.isel(time=ti, latitude=yi, longitude=xi).values
            out[i] = np.nan if (val is None) else float(val)

    # scale to fraction if values are in %
    with np.errstate(invalid="ignore"):
        if np.nanmax(out) > 1.5:
            out = out / 100.0
    return out
```

```python
# run function to load data 

def process_year(year):
    print(f"\n=== October {year} ===")
    idx = download_sprof_index()
    dfg = filter_region_time_chlabpp(idx, year)
    if dfg.empty:
        print("No Sprof rows with CHLA+BBP in this box/month."); return
    print(f"Profiles with CHLA+BBP (index hits): {len(dfg)}")

    times = dfg["datetime"].to_numpy()
    lats  = dfg["lat"].astype(float).to_numpy()
    lons  = dfg["lon"].astype(float).to_numpy()

    ice_vals = sample_oisst_ice_daily(times, lats, lons)
    print("Ice stats at points (min/max):", np.nanmin(ice_vals), np.nanmax(ice_vals))
    under = ice_vals >= ICE_THRESHOLD
    print(f"Under-ice ({ICE_THRESHOLD:.2f}): {int(np.nansum(under))} / {len(under)}")

    out = pd.DataFrame({
        "wmo_id": dfg["wmo_id"].astype(str),
        "cycle_number": pd.to_numeric(dfg["cycle_number"], errors="coerce"),
        "time": pd.to_datetime(times, utc=True),
        "latitude": lats, "longitude": lons,
        "ice_fraction": ice_vals, "under_ice": under
    })
    out.to_csv(os.path.join(OUTDIR, f"under_ice_{year}_october_weddell.csv"), index=False)
    print("Saved CSV.")

for yr in YEARS:
    try: process_year(yr)
    except Exception as e: print(f"Error {yr}: {e}")
print("\nDone.")
```

```python
# match each under-ice profile to nearest argo profile
# download the NetCDF, extract key bio-optical variables, compute layer-averaged values, + save a summary CSV

OUTDIR = "./poc_outputs"; os.makedirs(OUTDIR, exist_ok=True)
TMP_NC = os.path.join(OUTDIR, "sprof_netcdf"); os.makedirs(TMP_NC, exist_ok=True)

# rebuild 'mt' if not in memory:
try:
    mt
except NameError:
    # minimal rebuild + fresh index
    import io, gzip
    def load_sprof_index():
        for url in ["https://data-argo.ifremer.fr/argo_synthetic-profile_index.txt.gz",
                    "https://usgodae.org/ftp/outgoing/argo/argo_synthetic-profile_index.txt.gz"]:
            try:
                r = requests.get(url, timeout=120); r.raise_for_status()
                raw = gzip.decompress(r.content)
                names = ["file","date","lat","lon","ocean","profiler_type","institution","parameters","parameter_data_mode","date_update"]
                df = pd.read_csv(io.BytesIO(raw), comment="#", header=None, sep=",", names=names, dtype=str, engine="python")
                df["date"] = pd.to_datetime(df["date"], format="%Y%m%d%H%M%S", utc=True, errors="coerce")
                df["lat"]  = pd.to_numeric(df["lat"], errors="coerce")
                df["lon"]  = ((pd.to_numeric(df["lon"], errors="coerce")+180)%360)-180
                P = df["parameters"].fillna("").str.upper()
                df["has_chl"] = P.str.contains(r"\bCHL[A]?\b|CHLORO|CHLOROPHYLL", regex=True)
                df["has_bbp"] = P.str.contains(r"\bBBP\b|BBP\d{3}|BETA_BACKSCATTER", regex=True)
                df["has_both"] = df["has_chl"] & df["has_bbp"]
                return df
            except Exception:
                continue
        raise RuntimeError("index download failed")
    idx = load_sprof_index()

    def haversine_km(a,b,c,d):
        R=6371; import math
        p1,p2 = math.radians(a), math.radians(c)
        dphi  = math.radians(c-a); dl = math.radians(d-b)
        A = math.sin(dphi/2)**2 + math.cos(p1)*math.cos(p2)*math.sin(dl/2)**2
        return 2*R*math.asin(math.sqrt(A))
    def match_with_pref(r, pool):
        t0 = pd.to_datetime(r["time"], utc=True)
        cand = pool[(pool["date"]>=t0-pd.Timedelta("2D"))&(pool["date"]<=t0+pd.Timedelta("2D"))]
        cand = cand[(cand["lat"].between(r["latitude"]-0.7, r["latitude"]+0.7)) &
                    (cand["lon"].between(r["longitude"]-0.7, r["longitude"]+0.7))]
        def pick(sub):
            if sub.empty: return None
            sub = sub.copy()
            sub["dt_h"] = (sub["date"]-t0).abs().dt.total_seconds()/3600
            sub["dist"] = [haversine_km(r["latitude"], r["longitude"], la, lo) for la,lo in zip(sub["lat"], sub["lon"])]
            return sub.sort_values(by=["dist","dt_h"]).iloc[0]
        best = pick(cand[cand["has_both"]]) or pick(cand[cand["has_chl"]|cand["has_bbp"]]) or pick(cand)
        return best
    matched=[]
    for _,row in pts.iterrows():
        b=match_with_pref(row, idx)
        matched.append({**row.to_dict(),
                        "sprof_file": (b["file"] if b is not None else None),
                        "sprof_time": (b["date"] if b is not None else pd.NaT)})
    mt = pd.DataFrame(matched)

def fetch_netcdf(relpath, dest):
    rel = ensure_gdac_path(relpath)
    for base in ("https://data-argo.ifremer.fr/", "https://usgodae.org/ftp/outgoing/argo/"):
        try:
            rr = requests.get(base + rel, timeout=200)
            rr.raise_for_status()
            with open(dest, "wb") as f: f.write(rr.content)
            return dest
        except Exception:
            pass
    raise RuntimeError(f"Could not download: {rel}")

def pick_var(ds):
    # choose one CHL* and BBP*
    varU = {v.upper(): v for v in ds.data_vars}
    # chlorophyll
    for c in ("CHLA_ADJUSTED","CHLOROPHYLL_ADJUSTED","CHLA","CHLOROPHYLL"):
        if c in varU: chl = varU[c]; break
    else: chl=None
    # explicit wavelengths first for bbp
    bbp_candidates = [v for v in ds.data_vars if v.upper().startswith("BBP")]
    bbp_adj = [v for v in bbp_candidates if v.upper().endswith("_ADJUSTED")]
    bbp = bbp_adj[0] if bbp_adj else (bbp_candidates[0] if bbp_candidates else None)
    # also allow BETA_BACKSCATTER*
    if bbp is None:
        beta = [v for v in ds.data_vars if v.upper().startswith("BETA_BACKSCATTER")]
        beta_adj = [v for v in beta if v.upper().endswith("_ADJUSTED")]
        bbp = beta_adj[0] if beta_adj else (beta[0] if beta else None)
    # pressure
    pres = "PRES_ADJUSTED" if "PRES_ADJUSTED" in ds else ("PRES" if "PRES" in ds else None)
    return pres, chl, bbp

def qc_good_mask(ds, var):
    base = var.replace("_ADJUSTED","")
    for qn in (var+"_QC", base+"_QC"):
        if qn in ds.variables:
            qc = ds[qn].values
            if getattr(qc, "dtype", None) and qc.dtype.kind == "S":
                qc = qc.astype("U1")
            else:
                qc = np.array(qc).astype(str)
                qc = np.char.strip(qc)
                qc = np.char.replace(np.char.replace(qc, "b'", ""), "'", "")
            return np.isin(qc, ["1","2"])
    return np.ones(ds[var].shape, dtype=bool)

def mean_layer(p, x, m, z0, z1):
    p = np.asarray(p,float).ravel(); x = np.asarray(x,float).ravel(); m = np.asarray(m,bool).ravel()
    sel = np.isfinite(p) & np.isfinite(x) & m & (p>=z0) & (p<=z1)
    return float(np.nanmean(x[sel])) if np.any(sel) else np.nan

def top50_from_shallowest(p, x, m):
    p = np.asarray(p,float).ravel(); x = np.asarray(x,float).ravel(); m = np.asarray(m,bool).ravel()
    mm = np.isfinite(p) & np.isfinite(x) & m
    if not mm.any(): return np.nan
    z0 = float(np.nanmin(p[mm])); return mean_layer(p, x, m, z0, z0+50.0)

rows=[]
for _, r in mt.iterrows():
    rel = r["sprof_file"]
    local = os.path.join(TMP_NC, os.path.basename(rel))
    if not os.path.exists(local):
        try: fetch_netcdf(rel, local)
        except Exception: 
            rows.append({**r.to_dict(), "chla_top50obs":np.nan, "bbp_top50obs":np.nan,
                         "chla_50_250":np.nan, "bbp_50_250":np.nan})
            continue
    try:
        ds = xr.open_dataset(local)
        pres, chl, bbp = pick_var(ds)
        chla_top = bbp_top = chla_50_250 = bbp_50_250 = np.nan
        if pres and (chl or bbp):
            P = ds[pres].values
            if chl:
                C = ds[chl].values; Gc = qc_good_mask(ds, chl)
                chla_top    = top50_from_shallowest(P, C, Gc)
                chla_50_250 = mean_layer(P, C, Gc, 50.0, 250.0)
            if bbp:
                B = ds[bbp].values; Gb = qc_good_mask(ds, bbp)
                bbp_top     = top50_from_shallowest(P, B, Gb)
                bbp_50_250  = mean_layer(P, B, Gb, 50.0, 250.0)
        ds.close()
    except Exception:
        chla_top = bbp_top = chla_50_250 = bbp_50_250 = np.nan

    rows.append({**r.to_dict(),
                 "chla_top50obs": chla_top, "bbp_top50obs": bbp_top,
                 "chla_50_250": chla_50_250, "bbp_50_250": bbp_50_250})

v3 = pd.DataFrame(rows)
v3_path = os.path.join(OUTDIR, "under_ice_bio_summary_v3.csv")
v3.to_csv(v3_path, index=False)
print("Saved:", v3_path)
print(v3[["sprof_file","chla_top50obs","bbp_top50obs","chla_50_250","bbp_50_250"]].head(10))
```

```python
IN_CSV  = "./poc_outputs/under_ice_bio_summary_v3.csv"
OUT_DIR = "./poc_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# load table
tbl = pd.read_csv(IN_CSV, parse_dates=["time"])
print(f"Loaded {len(tbl)} rows from {IN_CSV}")

# under-ice transmittance
lam = np.arange(400, 701, 5)  # nm

def k_snow(l):
    # effective spectral attenuation for snow [m^-1]
    x = (l - 500.0)
    return 6.0 + 0.00005 * x**2 + 0.003 * (l - 500.0)

def k_ice(l):
    # effective spectral attenuation for sea ice [m^-1]
    x = (l - 500.0)
    return 1.5 + 0.00002 * x**2 + 0.0007 * (l - 500.0)

def transmittance_under_ice(h_snow, h_ice):
    Ks = k_snow(lam)
    Ki = k_ice(lam)
    return np.exp(-(Ks * h_snow + Ki * h_ice))

def blue_green_index(T, lam, blue=(440,500), green=(520,560)):
    B = np.nanmean(T[(lam >= blue[0]) & (lam <= blue[1])])
    G = np.nanmean(T[(lam >= green[0]) & (lam <= green[1])])
    return np.nan if (not np.isfinite(B) or not np.isfinite(G) or G == 0) else (B / G)

# Map ice_fraction  plausible (snow_m, ice_m)
def thickness_from_ice_fraction(ice_frac):
    z = float(ice_frac)
    if z >= 0.95:
        return 0.35, 1.50  # thick/compact ice
    elif z >= 0.90:
        return 0.25, 1.20  # medium ice
    else:
        return 0.15, 0.80  # thinner/looser ice

# compute BGI (blue-green index) per profile
BGI, snow_m, ice_m = [], [], []
for z in tbl["ice_fraction"].values:
    hs, hi = thickness_from_ice_fraction(z)
    T = transmittance_under_ice(hs, hi)
    BGI.append(blue_green_index(T, lam))
    snow_m.append(hs)
    ice_m.append(hi)

tbl["BGI_blue_green"] = BGI
tbl["snow_m"] = snow_m
tbl["ice_m"]  = ice_m

# small-particle proxy: bbp/chla (50250 m), log10
with np.errstate(divide="ignore", invalid="ignore"):
    ratio = tbl["bbp_50_250"].values / tbl["chla_50_250"].values
tbl["bbp_over_chla_50_250"] = ratio
tbl["log10_bbp_over_chla"]  = np.log10(ratio)

# correlations
try:
    from scipy.stats import pearsonr, spearmanr
    USE_SCIPY = True
except Exception:
    USE_SCIPY = False

m = np.isfinite(tbl["BGI_blue_green"].values) & np.isfinite(tbl["log10_bbp_over_chla"].values)
x = tbl.loc[m, "BGI_blue_green"].values
y = tbl.loc[m, "log10_bbp_over_chla"].values

if m.sum() >= 3:
    if USE_SCIPY:
        pr = pearsonr(x, y)  # statistic, pvalue
        sr = spearmanr(x, y)
        print(f"BGI vs log10(bbp/chla): Pearson r={pr.statistic:.2f} (p={pr.pvalue:.3g}), "
              f"Spearman ={sr.statistic:.2f} (p={sr.pvalue:.3g}), n={m.sum()}")
        r_for_title = pr.statistic
    else:
        r_for_title = float(np.corrcoef(x, y)[0,1])
        print(f"BGI vs log10(bbp/chla): Pearson r={r_for_title:.2f} (scipy not installed), n={m.sum()}")
else:
    r_for_title = np.nan
    print("Not enough finite pairs for correlation.")

# plots

# scatter: BGI vs log10(bbp/chla)
plt.figure()
plt.scatter(x, y)
plt.xlabel("Blue:Green Transmittance Index (BGI)")
plt.ylabel("log10( bbp/chla )  [50250 m]")
plt.title("Under-ice Weddell (Oct 2020 & 2021): Light vs size proxy")
if np.isfinite(r_for_title):
    plt.suptitle(f"Pearson r = {r_for_title:.2f}", y=0.98, fontsize=10)
plt.tight_layout()
scatter_path = os.path.join(OUT_DIR, "bgi_vs_log10_bbp_over_chla.png")
plt.savefig(scatter_path, dpi=200)
plt.show()
print("Saved:", scatter_path)

# sanity checks
print("BGI range:", float(np.nanmin(tbl["BGI_blue_green"])), "", float(np.nanmax(tbl["BGI_blue_green"])))
print("Snow (m)  min/median/max:", np.nanmin(tbl["snow_m"]), np.nanmedian(tbl["snow_m"]), np.nanmax(tbl["snow_m"]))
print("Ice  (m)  min/median/max:",  np.nanmin(tbl["ice_m"]),  np.nanmedian(tbl["ice_m"]),  np.nanmax(tbl["ice_m"]))

plt.figure()
vals = tbl["BGI_blue_green"].dropna().values
plt.hist(vals, bins=min(10, max(5, int(np.sqrt(vals.size))) ))
plt.xlabel("BGI"); plt.ylabel("count")
plt.title("Blue:Green Index distribution")
plt.tight_layout()
hist_path = os.path.join(OUT_DIR, "hist_BGI.png")
plt.savefig(hist_path, dpi=200)
plt.show()
print("Saved:", hist_path)

# thickness sensitivity tests
def bgi_from_fixed(hs, hi):
    T = transmittance_under_ice(hs, hi)
    return blue_green_index(T, lam)

def bgi_from_scaled(scale=1.0):
    out = []
    for z in tbl["ice_fraction"].values:
        hs, hi = thickness_from_ice_fraction(z)
        T = transmittance_under_ice(scale*hs, scale*hi)
        out.append(blue_green_index(T, lam))
    return np.array(out)

scenarios = {
    "fixed_thin":   (0.10, 0.60),
    "fixed_medium": (0.25, 1.20),
    "fixed_thick":  (0.40, 1.80),
    "map_x0.7":     "scale0.7",
    "map_x1.3":     "scale1.3",
}

rows = []
Y = tbl["log10_bbp_over_chla"].values

for name, spec in scenarios.items():
    if isinstance(spec, tuple):
        bgi = np.full(len(tbl), bgi_from_fixed(*spec))
    else:
        fac = float(spec.replace("scale",""))
        bgi = bgi_from_scaled(fac)

    mm = np.isfinite(bgi) & np.isfinite(Y)
    if mm.sum() >= 3:
        if USE_SCIPY:
            pr = pearsonr(bgi[mm], Y[mm]); sr = spearmanr(bgi[mm], Y[mm])
            rows.append({"scenario": name, "n": int(mm.sum()),
                         "pearson_r": float(pr.statistic), "pearson_p": float(pr.pvalue),
                         "spearman_rho": float(sr.statistic), "spearman_p": float(sr.pvalue)})
        else:
            r_np = float(np.corrcoef(bgi[mm], Y[mm])[0,1])
            rows.append({"scenario": name, "n": int(mm.sum()),
                         "pearson_r": r_np, "pearson_p": np.nan,
                         "spearman_rho": np.nan, "spearman_p": np.nan})
    else:
        rows.append({"scenario": name, "n": int(mm.sum()),
                     "pearson_r": np.nan, "pearson_p": np.nan,
                     "spearman_rho": np.nan, "spearman_p": np.nan})

sens = pd.DataFrame(rows)
print("\nSensitivity correlations:\n", sens)
sens_path = os.path.join(OUT_DIR, "bgi_sensitivity_correlations.csv")
sens.to_csv(sens_path, index=False)
print("Saved:", sens_path)

# save table
OUT_CSV = os.path.join(OUT_DIR, "under_ice_bio_with_BGI.csv")
tbl.to_csv(OUT_CSV, index=False)
print("Saved:", OUT_CSV)
```

```python
# compute spectral width metrics

OUT_DIR = "./poc_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# load table
path_with_bgi = os.path.join(OUT_DIR, "under_ice_bio_with_BGI.csv")
path_v3       = os.path.join(OUT_DIR, "under_ice_bio_summary_v3.csv")

if os.path.exists(path_with_bgi):
    tbl = pd.read_csv(path_with_bgi, parse_dates=["time"])
else:
    tbl = pd.read_csv(path_v3, parse_dates=["time"])
    # add snow_m/ice_m and bbp_over_chla/log10 if missing
    def thickness_from_ice_fraction(ice_frac):
        z = float(ice_frac)
        if z >= 0.95:   return 0.35, 1.50
        elif z >= 0.90: return 0.25, 1.20
        else:           return 0.15, 0.80
    if "snow_m" not in tbl.columns or "ice_m" not in tbl.columns:
        snow_m, ice_m = [], []
        for z in tbl["ice_fraction"].values:
            hs, hi = thickness_from_ice_fraction(z)
            snow_m.append(hs); ice_m.append(hi)
        tbl["snow_m"] = snow_m
        tbl["ice_m"]  = ice_m
    if "bbp_over_chla_50_250" not in tbl.columns:
        with np.errstate(divide="ignore", invalid="ignore"):
            ratio = tbl["bbp_50_250"].values / tbl["chla_50_250"].values
        tbl["bbp_over_chla_50_250"] = ratio
        tbl["log10_bbp_over_chla"]  = np.log10(ratio)
    # compute BGI if missing
    if "BGI_blue_green" not in tbl.columns:
        print("Note: BGI not found; computing with default model so we can plot width vs BGI.")
        # spectral model copied from the BGI cell:
        lam_tmp = np.arange(400, 701, 5)
        def k_snow(l): x=(l-500.0); return 6.0 + 0.00005*x**2 + 0.003*(l-500.0)
        def k_ice(l):  x=(l-500.0); return 1.5 + 0.00002*x**2 + 0.0007*(l-500.0)
        def T_under_ice(hs,hi):
            Ks, Ki = k_snow(lam_tmp), k_ice(lam_tmp)
            return np.exp(-(Ks*hs + Ki*hi))
        def blue_green_index(T, lam, blue=(440,500), green=(520,560)):
            B = np.nanmean(T[(lam>=blue[0])&(lam<=blue[1])])
            G = np.nanmean(T[(lam>=green[0])&(lam<=green[1])])
            return np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
        bgi=[]
        for hs,hi in zip(tbl["snow_m"].values, tbl["ice_m"].values):
            bgi.append(blue_green_index(T_under_ice(hs,hi), lam_tmp))
        tbl["BGI_blue_green"] = bgi

print(f"Loaded {len(tbl)} rows")

# define the same under-ice spectral model used for BGI
lam = np.arange(400, 701, 5)  # wavelength grid [nm]

def k_snow(l):
    x = (l - 500.0)
    return 6.0 + 0.00005 * x**2 + 0.003 * (l - 500.0)

def k_ice(l):
    x = (l - 500.0)
    return 1.5 + 0.00002 * x**2 + 0.0007 * (l - 500.0)

def transmittance_under_ice(h_snow, h_ice):
    Ks = k_snow(lam)
    Ki = k_ice(lam)
    return np.exp(-(Ks * h_snow + Ki * h_ice))

# spectral-shape metrics helpers
def spectral_metrics(T, lam):
    """
    T : array of transmittance at lam (same length)
    Returns dict with:
      - fwhm_nm: full width at half maximum (nm)
      - centroid_nm: spectral centroid (nm), T-weighted
      - stdev_nm: spectral standard deviation (nm)
      - entropy_norm: normalized Shannon entropy in [0,1]
      - max_T: peak value of T
      - lambda_at_max: wavelength of peak T
    """
    T = np.asarray(T, float).copy()
    lam = np.asarray(lam, float)
    # guard
    if T.size < 3 or not np.isfinite(T).any():
        return dict(fwhm_nm=np.nan, centroid_nm=np.nan, stdev_nm=np.nan,
                    entropy_norm=np.nan, max_T=np.nan, lambda_at_max=np.nan)

    # normalize for FWHM
    Tmax = np.nanmax(T)
    if not np.isfinite(Tmax) or Tmax <= 0:
        fwhm = np.nan
    else:
        Tn = T / Tmax
        # half-max mask; pick the main contiguous region around the global max
        idx_max = int(np.nanargmax(Tn))
        mask = Tn >= 0.5
        # find contiguous block around idx_max
        # expand left
        i0 = idx_max
        while i0-1 >= 0 and mask[i0-1]:
            i0 -= 1
        # expand right
        i1 = idx_max
        while i1+1 < len(mask) and mask[i1+1]:
            i1 += 1
        # linear-edge refinement 
        fwhm = lam[i1] - lam[i0]

    # centroid & spread using T as weights
    Tw = np.clip(T, 0, np.inf)
    S = np.nansum(Tw)
    if not np.isfinite(S) or S <= 0:
        centroid = np.nan; stdev = np.nan
    else:
        centroid = float(np.nansum(lam * Tw) / S)
        var = float(np.nansum(((lam - centroid) ** 2) * Tw) / S)
        stdev = np.sqrt(var)

    # normalized spectral entropy (0..1)
    if not np.isfinite(S) or S <= 0:
        Hn = np.nan
    else:
        p = Tw / S
        # avoid log(0)
        with np.errstate(divide="ignore", invalid="ignore"):
            H = -np.nansum(np.where(p > 0, p * np.log(p), 0.0))
        Hmax = np.log(len(p))
        Hn = float(H / Hmax) if Hmax > 0 else np.nan

    lam_peak = float(lam[np.nanargmax(T)]) if np.isfinite(Tmax) else np.nan

    return dict(fwhm_nm=float(fwhm),
                centroid_nm=float(centroid),
                stdev_nm=float(stdev),
                entropy_norm=float(Hn),
                max_T=float(Tmax),
                lambda_at_max=float(lam_peak))

# compute metrics per profile
rows = []
for i, r in tbl.iterrows():
    hs = float(r["snow_m"])
    hi = float(r["ice_m"])
    T = transmittance_under_ice(hs, hi)
    met = spectral_metrics(T, lam)
    rows.append({
        "sprof_file": r.get("sprof_file", ""),
        "time": r.get("time", ""),
        "latitude": r.get("latitude", np.nan),
        "longitude": r.get("longitude", np.nan),
        "ice_fraction": r.get("ice_fraction", np.nan),
        "snow_m": hs, "ice_m": hi,
        "BGI_blue_green": r.get("BGI_blue_green", np.nan),
        "bbp_50_250": r.get("bbp_50_250", np.nan),
        "chla_50_250": r.get("chla_50_250", np.nan),
        "bbp_over_chla_50_250": r.get("bbp_over_chla_50_250", np.nan),
        "log10_bbp_over_chla": r.get("log10_bbp_over_chla", np.nan),
        **met
    })

spec = pd.DataFrame(rows)
out_csv = os.path.join(OUT_DIR, "under_ice_spectral_shape_metrics.csv")
spec.to_csv(out_csv, index=False)
print("Saved:", out_csv)
print(spec.head(5)[["sprof_file","fwhm_nm","centroid_nm","entropy_norm","BGI_blue_green"]])

# quick sum-up
def qstats(x):
    x = np.asarray(x, float); x = x[np.isfinite(x)]
    if x.size == 0: return "n=0"
    p25, p50, p75 = np.percentile(x, [25,50,75])
    return f"n={x.size}, median={p50:.1f}, IQR=({p25:.1f}, {p75:.1f})"

print("FWHM (nm):", qstats(spec["fwhm_nm"]))
print("Centroid  (nm):", qstats(spec["centroid_nm"]))
print("Entropy (norm):", qstats(spec["entropy_norm"]))

# plots

# histogram of FWHM
plt.figure()
vals = spec["fwhm_nm"].dropna().values
plt.hist(vals, bins=min(10, max(5, int(np.sqrt(vals.size)))))
plt.xlabel("Spectral FWHM (nm)")
plt.ylabel("count")
plt.title("Under-ice spectral width (FWHM)")
plt.tight_layout()
p1 = os.path.join(OUT_DIR, "hist_spectral_fwhm_nm.png")
plt.savefig(p1, dpi=200)
plt.show()
print("Saved:", p1)

# histogram of centroid wavelength
plt.figure()
vals = spec["centroid_nm"].dropna().values
plt.hist(vals, bins=min(10, max(5, int(np.sqrt(vals.size)))))
plt.xlabel("Spectral centroid (nm)")
plt.ylabel("count")
plt.title("Under-ice spectral centroid")
plt.tight_layout()
p2 = os.path.join(OUT_DIR, "hist_spectral_centroid_nm.png")
plt.savefig(p2, dpi=200)
plt.show()
print("Saved:", p2)

# scatter: width vs BGI 
if "BGI_blue_green" in spec.columns and np.isfinite(spec["BGI_blue_green"]).any():
    m = np.isfinite(spec["fwhm_nm"]) & np.isfinite(spec["BGI_blue_green"])
    plt.figure()
    plt.scatter(spec.loc[m,"BGI_blue_green"], spec.loc[m,"fwhm_nm"])
    plt.xlabel("Blue:Green Index (BGI)")
    plt.ylabel("FWHM (nm)")
    plt.title("Spectral width vs under-ice color balance")
    plt.tight_layout()
    p3 = os.path.join(OUT_DIR, "scatter_width_vs_bgi.png")
    plt.savefig(p3, dpi=200); plt.show()
    print("Saved:", p3)

# scatter: width vs small-particle proxy
if "log10_bbp_over_chla" in spec.columns and np.isfinite(spec["log10_bbp_over_chla"]).any():
    m = np.isfinite(spec["fwhm_nm"]) & np.isfinite(spec["log10_bbp_over_chla"])
    plt.figure()
    plt.scatter(spec.loc[m,"fwhm_nm"], spec.loc[m,"log10_bbp_over_chla"])
    plt.xlabel("FWHM (nm)")
    plt.ylabel("log10(bbp/chla) [50250 m]")
    plt.title("Spectral width vs size proxy")
    plt.tight_layout()
    p4 = os.path.join(OUT_DIR, "scatter_width_vs_log10_bbp_over_chla.png")
    plt.savefig(p4, dpi=200); plt.show()
    print("Saved:", p4)
```

```python
# poc using figshare files (irradiance + a_ph())
# loads table, reads *Antarctica* under-ice irradiance spectra from figshare
# reads phytoplankton absorption spectra files (cryptophytes, ice diatoms, P. antarctica, prasinophytes)
# computes BGI, FWHM, centroid for each irradiance column
# - Maps profiles to thin / medium / thick (by ice_fraction), then assigns:
#       thin   bluest/narrowest spectrum (highest BGI, lowest centroid)
#       medium middle spectrum
#       thick  greenest/broadest spectrum (lowest BGI, highest centroid)
# - Computes PUR =  E() * a_ph() *  per phytoplankton group for each profile
# - Re-plots BGI vs log10(bbp/chla) and saves merged outputs

POC_DIR        = "./sojawozniak"  
ABS_DIR        = "./sojawozniak"       
IRR_DIR        = "./sojawozniak"       
USE_SIMULATED  = True 
ABS_FILES = [
    "Cryptophytes_absortpion.csv",
    "ice_diatoms_absorption.csv",           
    "Phaeocystis_antarctica _absortpion.csv",
    "Prasinophyte_absortpion.csv"
]
IRR_FILE = ("simulated_irradiance_spectra_Antarctica.csv" 
            if USE_SIMULATED else 
            "measured_irradiance_spectra_Antarctica.csv")

# helper functions 
def load_profiles():
    p_bgi = os.path.join(POC_DIR, "under_ice_bio_with_BGI.csv")
    p_v3  = os.path.join(POC_DIR, "under_ice_bio_summary_v3.csv")
    if os.path.exists(p_bgi):
        df = pd.read_csv(p_bgi, parse_dates=["time"])
    else:
        df = pd.read_csv(p_v3, parse_dates=["time"])
    if "bbp_over_chla_50_250" not in df.columns:
        with np.errstate(divide="ignore", invalid="ignore"):
            ratio = df["bbp_50_250"].values / df["chla_50_250"].values
        df["bbp_over_chla_50_250"] = ratio
        df["log10_bbp_over_chla"]  = np.log10(ratio)
    return df

def ice_bin(frac):
    z = float(frac)
    # within "under-ice" (0.15), use relative bins
    if z >= 0.95:   return "thick"
    elif z >= 0.90: return "medium"
    else:           return "thin"

def col_is_wavelength(name):
    n = str(name).strip().lower()
    return any(k in n for k in ["wavelength","lambda","nm","wl","wave"])

def pick_wl_column(df):
    for i, c in enumerate(df.columns):
        if col_is_wavelength(c):
            return i
    best = None
    for i in range(df.shape[1]):
        v = pd.to_numeric(df.iloc[:, i], errors="coerce")
        if v.notna().sum() > 10 and v.between(350, 750).mean() > 0.5:
            best = i; break
    return 0 if best is None else best

def read_spectra_table(path):
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam >= 380) & (lam <= 750)
    lam = lam[m]
    spectra = {}
    for j in range(df.shape[1]):
        if j == wl_idx: 
            continue
        y = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(y).sum() >= 10:
            name = str(df.columns[j]).strip()
            spectra[name] = y
    if not spectra:
        raise ValueError(f"No valid spectra columns in {os.path.basename(path)}")
    return lam, spectra

def metrics_for(lam, S):
    S = np.asarray(S, float)
    # BGI
    B = np.nanmean(S[(lam>=440)&(lam<=500)])
    G = np.nanmean(S[(lam>=520)&(lam<=560)])
    BGI = np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
    # FWHM
    Tmax = np.nanmax(S)
    if not np.isfinite(Tmax) or Tmax <= 0:
        FWHM = np.nan
    else:
        Tn = S / Tmax
        kmax = int(np.nanargmax(Tn))
        mask = Tn >= 0.5
        i0 = kmax
        while i0-1>=0 and mask[i0-1]: i0 -= 1
        i1 = kmax
        while i1+1<len(mask) and mask[i1+1]: i1 += 1
        FWHM = float(lam[i1] - lam[i0])
    # centroid (shape-weighted mean )
    Spos = np.clip(S, 0, np.inf); Ssum = np.nansum(Spos)
    centroid = float(np.nansum(lam * Spos) / Ssum) if Ssum > 0 else np.nan
    return float(BGI), float(FWHM), float(centroid)

def rank_irradiance_spectra(lam, spectra_dict):
    rows = []
    for name, S in spectra_dict.items():
        BGI, FWHM, CEN = metrics_for(lam, S)
        rows.append({"col": name, "BGI": BGI, "FWHM": FWHM, "centroid": CEN})
    df = pd.DataFrame(rows)
    # bluish: high BGI, low centroid; greenish: low BGI, high centroid
    df = df.sort_values(["BGI","centroid"], ascending=[False, True]).reset_index(drop=True)
    return df

def read_absorption_file(path):
    # each file should have wavelength + absorption()
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    # find first non-wavelength numeric column
    A = None
    label = None
    for j in range(df.shape[1]):
        if j == wl_idx: continue
        v = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(v).sum() >= 10:
            A = v; label = str(df.columns[j]).strip(); break
    if A is None:
        raise ValueError(f"No numeric absorption column in {os.path.basename(path)}")
    # ensure non-negative and smooth
    A = np.clip(A, 0, np.inf)
    return lam, A, label

def resample_to(lam_src, y_src, lam_dst):
    # simple linear interpolation onto lam_dst
    y = np.interp(lam_dst, lam_src, y_src, left=np.nan, right=np.nan)
    # fill small NaN edges by nearest
    if np.isnan(y[0]):  y[0]  = y[~np.isnan(y)][0]
    if np.isnan(y[-1]): y[-1] = y[~np.isnan(y)][-1]
    return y

def compute_PUR(lam_nm, E, A):
    # integrate E() * a_ph() over  (nm); unit scaling cancels in ratios
    # : assume uniform nm spacing
    dlam = np.median(np.diff(lam_nm))
    return float(np.nansum(E * A) * dlam)

# load profiles
tbl = load_profiles()
tbl["ice_bin"] = [ice_bin(z) for z in tbl["ice_fraction"]]

# irradiance spectra
irr_path = os.path.join(IRR_DIR, IRR_FILE)
if not os.path.exists(irr_path):
    raise FileNotFoundError(f"Irradiance file not found: {irr_path}")
lamE, Ecols = read_spectra_table(irr_path)
ranked = rank_irradiance_spectra(lamE, Ecols)
print("Irradiance spectra ranked (bluishgreenish):")
print(ranked.head(10))

# choose representatives
if len(ranked) == 1:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[0]["col"]}
elif len(ranked) == 2:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[1]["col"]}
else:
    reps = {
        "thin":   ranked.iloc[0]["col"],                   # highest BGI / bluest
        "medium": ranked.iloc[len(ranked)//2]["col"],      # middle
        "thick":  ranked.iloc[-1]["col"],                  # lowest BGI / greenest
    }
print("\nRepresentative spectra:", reps)

# assign spectrum to each profile
tbl["irr_col"] = [reps[b] for b in tbl["ice_bin"]]
# compute per-profile BGI/FWHM/centroid from assigned spectrum
BGI_list, FWHM_list, CEN_list = [], [], []
for col in tbl["irr_col"]:
    BGI, FWHM, CEN = metrics_for(lamE, Ecols[col])
    BGI_list.append(BGI); FWHM_list.append(FWHM); CEN_list.append(CEN)
tbl["BGI_ED"]         = BGI_list
tbl["FWHM_nm_ED"]     = FWHM_list
tbl["centroid_nm_ED"] = CEN_list

# absorption spectra (4 groups)
abs_files_found = []
for f in ABS_FILES:
    candidates = [fn for fn in os.listdir(ABS_DIR) if fn.replace(" ","").lower()==f.replace(" ","").lower()]
    if not candidates:
        # try loose contains
        stem = re.sub(r"[^a-z]", "", f.lower())
        for fn in os.listdir(ABS_DIR):
            if stem in re.sub(r"[^a-z]","", fn.lower()):
                candidates.append(fn)
    if not candidates:
        raise FileNotFoundError(f"Absorption file not found (tried variants): {f}")
    abs_files_found.append(os.path.join(ABS_DIR, candidates[0]))

A_dict = {} 
for p in abs_files_found:
    lamA, A, label = read_absorption_file(p)
    # resample to irradiance wavelength grid
    A_res = resample_to(lamA, A, lamE)
    key = os.path.splitext(os.path.basename(p))[0]
    A_dict[key] = A_res

print("\nAbsorption spectra loaded:", list(A_dict.keys()))

# PUR per profile  phytoplankton group 
for key, A in A_dict.items():
    pur_vals = []
    for col in tbl["irr_col"]:
        E = Ecols[col]
        pur_vals.append(compute_PUR(lamE, E, A))
    tbl[f"PUR_{key}"] = pur_vals

# normalize PUR by its median across profiles for readability
for key in A_dict.keys():
    med = np.nanmedian(tbl[f"PUR_{key}"].values)
    if np.isfinite(med) and med > 0:
        tbl[f"PUR_{key}_norm"] = tbl[f"PUR_{key}"] / med

# plots
from math import isnan
def safe_corr(x, y):
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() < 3: return np.nan, np.nan
    try:
        from scipy.stats import pearsonr
        r, p = pearsonr(x[m], y[m]); return float(r), float(p)
    except Exception:
        r = float(np.corrcoef(x[m], y[m])[0,1]); return r, np.nan

# BGI vs size proxy
r, p = safe_corr(tbl["BGI_ED"].values, tbl["log10_bbp_over_chla"].values)
print(f"\nBGI (assigned) vs log10(bbp/chla): r={r:.2f}, p={p if not isnan(p) else np.nan}, n={np.isfinite(tbl['log10_bbp_over_chla']).sum()}")

# PUR vs size proxy (for each phytoplankton group)
for key in A_dict.keys():
    rP, pP = safe_corr(tbl[f"PUR_{key}_norm"].values, tbl["log10_bbp_over_chla"].values)
    print(f"PUR_{key} (norm) vs log10(bbp/chla): r={rP:.2f}, p={pP if not isnan(pP) else np.nan}")

# save table
out_csv = os.path.join(POC_DIR, "poc_with_figshare_ED_irradiance_and_absorption.csv")
tbl.to_csv(out_csv, index=False)
print("\nSaved:", out_csv)

# plot: BGI vs size proxy
plt.figure()
plt.scatter(tbl["BGI_ED"], tbl["log10_bbp_over_chla"])
plt.xlabel("BGI (assigned from Antarctic irradiance spectra)")
plt.ylabel("log10(bbp/chla) [50250 m]")
plt.title("Under-ice Weddell (Oct 2020 & 2021): Light color vs size proxy")
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_BGI_vs_size.png"), dpi=200)
plt.show()

# plot: PUR (per group) vs size proxy
plt.figure()
for key in A_dict.keys():
    plt.scatter(tbl[f"PUR_{key}_norm"], tbl["log10_bbp_over_chla"], label=key)
plt.xlabel("PUR (normalized within group)")
plt.ylabel("log10(bbp/chla)")
plt.title("Under-ice: PUR vs size proxy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_PUR_vs_size.png"), dpi=200)
plt.show()

# plot: representatives spectral shape
reps_df = ranked.copy()
rep_names = [reps["thin"], reps["medium"], reps["thick"]]
plt.figure()
for nm in rep_names:
    plt.plot(lamE, Ecols[nm], label=f"{nm}")
plt.xlabel("Wavelength (nm)")
plt.ylabel("E() (arbitrary units)")
plt.title("Chosen irradiance spectra (thin=bluest, thick=greenest/broadest)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_representative_spectra.png"), dpi=200)
plt.show()
```

```python
# thickness-aware mapping

SPECTRA_DIR = "./sojawozniak" 

OUT_DIR = "./poc_outputs"; os.makedirs(OUT_DIR, exist_ok=True)
PATH_V3  = os.path.join(OUT_DIR, "under_ice_bio_summary_v3.csv")
PATH_BGI = os.path.join(OUT_DIR, "under_ice_bio_with_BGI.csv")

# load table
tbl = pd.read_csv(PATH_BGI if os.path.exists(PATH_BGI) else PATH_V3, parse_dates=["time"])

# utilities
def size_proxy_columns(df):
    if "bbp_over_chla_50_250" not in df.columns:
        with np.errstate(divide="ignore", invalid="ignore"):
            ratio = df["bbp_50_250"].values / df["chla_50_250"].values
        df["bbp_over_chla_50_250"] = ratio
        df["log10_bbp_over_chla"]  = np.log10(ratio)
    return df

def icefrac_bin(z):
    z = float(z)
    if z >= 0.95:   return "thick"
    elif z >= 0.90: return "medium"
    else:           return "thin"

def read_csv_any(path):
    return pd.read_csv(path)

def compute_metrics(lam, S):
    S = np.asarray(S, float)
    B = np.nanmean(S[(lam>=440)&(lam<=500)])
    G = np.nanmean(S[(lam>=520)&(lam<=560)])
    BGI = np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
    Tmax = np.nanmax(S)
    if not np.isfinite(Tmax) or Tmax<=0:
        FWHM = np.nan
    else:
        Tn = S/Tmax; kmax = int(np.nanargmax(Tn))
        mask = Tn >= 0.5
        i0=kmax
        while i0-1>=0 and mask[i0-1]: i0-=1
        i1=kmax
        while i1+1<len(mask) and mask[i1+1]: i1+=1
        FWHM = lam[i1]-lam[i0]
    Spos = np.clip(S,0,np.inf); Ssum = np.nansum(Spos)
    centroid = float(np.nansum(lam*Spos)/Ssum) if Ssum>0 else np.nan
    return float(BGI), float(FWHM), float(centroid)

def read_spectra_table(path):
    df = read_csv_any(path)
    # wavelength column
    cols = [str(c).strip().lower() for c in df.columns]
    wl_idx = None
    for i,c in enumerate(cols):
        if any(w in c for w in ("wavelength","lambda","nm","wl","wave")):
            wl_idx = i; break
    if wl_idx is None:
        # guess by range
        cand = []
        for i in range(df.shape[1]):
            try:
                v = pd.to_numeric(df.iloc[:,i], errors="coerce")
                if v.notna().sum()>10 and v.between(350,750).mean()>0.6:
                    cand.append(i)
            except Exception: pass
        wl_idx = cand[0] if cand else 0
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    spectra = {}
    for j in range(df.shape[1]):
        if j==wl_idx: continue
        v = pd.to_numeric(df.iloc[:, j], errors="coerce").values
        v = v[m]
        if np.isfinite(v).sum()>=10:
            spectra[str(df.columns[j]).strip()] = v
    if not spectra:
        raise ValueError(f"No numeric spectra columns in {os.path.basename(path)}")
    return lam, spectra

# locate thickness-encoded spectra
def likely_thickness_file(path):
    t = os.path.basename(path).lower()
    # look for snow/ice thickness hints and NOT "physical_depth" / "optical_depth"
    has_thickness = any(k in t for k in ["snow", "ice", "thick", "hsnow", "hice"])
    is_depth = any(k in t for k in ["physical_depth", "optical_depth", "at8physical", "at6optical"])
    return has_thickness and ("under" in t and "ice" in t) and not is_depth

candidate_files = []
for root, _, files in os.walk(SPECTRA_DIR):
    for fn in files:
        if fn.lower().endswith(".csv"):
            p = os.path.join(root, fn)
            if likely_thickness_file(p):
                candidate_files.append(p)

if not candidate_files:
    print("No thickness-encoded CSV found. Falling back to 'under-ice at 8 physical depths' with fixed mapping.")
    # locate the depths file
    depths_file = None
    for root, _, files in os.walk(SPECTRA_DIR):
        for fn in files:
            if fn.lower().endswith(".csv") and \
               "under" in fn.lower() and "ice" in fn.lower() and "8" in fn and "physical" in fn.lower():
                depths_file = os.path.join(root, fn); break
        if depths_file: break
    if depths_file is None:
        raise RuntimeError("Could not locate 'scalar_irradiance_undericecover_at8physical_depths.csv' in SPECTRA_DIR.")
    print("Using:", os.path.basename(depths_file))

    lam, spectra_dict = read_spectra_table(depths_file)
    # Rank columns by BGI (to understand which are blue/green)
    ranked = []
    for name, S in spectra_dict.items():
        BGI, FWHM, CENT = compute_metrics(lam, S)
        ranked.append((name, BGI, FWHM, CENT))
    ranked = pd.DataFrame(ranked, columns=["col","BGI","FWHM","centroid"]).sort_values("BGI", ascending=False).reset_index(drop=True)
    print("\nRanked (BGI highlow):\n", ranked)
    reps = {
        "thin":   ranked.iloc[-1]["col"],                 # greenest
        "medium": ranked.iloc[len(ranked)//2]["col"],     # middle
        "thick":  ranked.iloc[0]["col"],                  # bluest
    }
    print("\nUsing reps (thin/medium/thick):", reps)

    # assign per profile
    tbl = size_proxy_columns(tbl)
    tbl["bin_icefrac"] = [icefrac_bin(z) for z in tbl["ice_fraction"].values]
    tbl["SJZ_col"] = [reps[b] for b in tbl["bin_icefrac"].values]

    BGI_list, FWHM_list, CENT_list = [], [], []
    for col in tbl["SJZ_col"].values:
        S = spectra_dict[col]
        BGI, FWHM, CENT = compute_metrics(lam, S)
        BGI_list.append(BGI); FWHM_list.append(FWHM); CENT_list.append(CENT)
    tbl["BGI_SJZ"]         = BGI_list
    tbl["FWHM_nm_SJZ"]     = FWHM_list
    tbl["centroid_nm_SJZ"] = CENT_list

else:
    print("Found thickness-encoded CSV(s):")
    for p in candidate_files: print(" -", os.path.basename(p))

    spec_path = candidate_files[0]
    print("Using:", os.path.basename(spec_path))
    lam, spectra_dict = read_spectra_table(spec_path)

    # parse (snow, ice) from column names if present
    def parse_thick_from_colname(name):
        t = name.lower()
        m1 = re.search(r"snow[_\- ]?(\d+(\.\d+)?)(cm|m)", t)
        m2 = re.search(r"ice[_\- ]?(\d+(\.\d+)?)(cm|m)",  t)
        def _val(m):
            if not m: return None
            x = float(m.group(1)); unit = m.group(3)
            return x/100.0 if unit=="cm" else x
        return _val(m1), _val(m2)

    cols_info = []
    for name, S in spectra_dict.items():
        BGI, FWHM, CENT = compute_metrics(lam, S)
        s, i = parse_thick_from_colname(name)
        cols_info.append({"col": name, "snow_m": s, "ice_m": i, "BGI": BGI, "FWHM": FWHM, "centroid": CENT})
    ci = pd.DataFrame(cols_info)
    print("\nColumns summary:\n", ci.head(10))

    # if thickness parsed, map bins to nearest (snow,ice); else use rank by BGI
    has_thick = ci["snow_m"].notna().any() or ci["ice_m"].notna().any()
    tbl = size_proxy_columns(tbl)
    tbl["bin_icefrac"] = [icefrac_bin(z) for z in tbl["ice_fraction"].values]

    if has_thick:
        # choose representatives per bin by nearest nominal thickness:
        # thin(0.15,0.8), medium(0.25,1.2), thick(0.35,1.5)
        targets = {"thin":(0.15,0.80), "medium":(0.25,1.20), "thick":(0.35,1.50)}
        reps = {}
        for k,(ts,ti) in targets.items():
            # distance in (snow, ice) space
            d = (ci["snow_m"].fillna(ts)-ts)**2 + (ci["ice_m"].fillna(ti)-ti)**2
            reps[k] = ci.iloc[int(np.argmin(d.values))]["col"]
        print("\nUsing reps (from parsed thickness):", reps)
    else:
        ranked = ci.sort_values("BGI", ascending=False).reset_index(drop=True)
        reps = {"thin": ranked.iloc[-1]["col"], "medium": ranked.iloc[len(ranked)//2]["col"], "thick": ranked.iloc[0]["col"]}
        print("\nUsing reps by BGI rank (no thickness labels in columns):", reps)

    tbl["SJZ_col"] = [reps[b] for b in tbl["bin_icefrac"].values]
    BGI_list, FWHM_list, CENT_list = [], [], []
    for col in tbl["SJZ_col"].values:
        S = spectra_dict[col]
        BGI, FWHM, CENT = compute_metrics(lam, S)
        BGI_list.append(BGI); FWHM_list.append(FWHM); CENT_list.append(CENT)
    tbl["BGI_SJZ"]         = BGI_list
    tbl["FWHM_nm_SJZ"]     = FWHM_list
    tbl["centroid_nm_SJZ"] = CENT_list

out_csv = os.path.join(OUT_DIR, "under_ice_bio_with_SJZ_thicknessaware.csv")
tbl.to_csv(out_csv, index=False); print("\nSaved:", out_csv)

print("\nAssigned spectra counts:")
print(tbl["SJZ_col"].value_counts())

# correlation with size proxy
m = np.isfinite(tbl["BGI_SJZ"].values) & np.isfinite(tbl["log10_bbp_over_chla"].values)
x = tbl.loc[m,"BGI_SJZ"].values; y = tbl.loc[m,"log10_bbp_over_chla"].values
try:
    from scipy.stats import pearsonr, spearmanr
    if m.sum()>=3:
        pr = pearsonr(x,y); sr = spearmanr(x,y)
        print(f"\nSJZ(BGI) vs log10(bbp/chla): r={pr.statistic:.2f} (p={pr.pvalue:.3g}), "
              f"rho={sr.statistic:.2f} (p={sr.pvalue:.3g}), n={m.sum()}")
except Exception:
    if m.sum()>=3:
        r = float(np.corrcoef(x,y)[0,1]); print(f"\nPearson r  {r:.2f} (scipy not installed), n={m.sum()}")

# plots
plt.figure()
plt.scatter(tbl["BGI_SJZ"], tbl["log10_bbp_over_chla"])
plt.xlabel("BGI (SojaWoniak; thickness-aware)")
plt.ylabel("log10(bbp/chla) [50250 m]")
plt.title("Under-ice Weddell: Light vs size proxy (thickness-aware)")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR,"bgi_vs_size_SJZ_thicknessaware.png"), dpi=200); plt.show()

plt.figure()
plt.scatter(tbl["centroid_nm_SJZ"], tbl["FWHM_nm_SJZ"])
plt.xlabel("Spectral centroid (nm)")
plt.ylabel("FWHM (nm)")
plt.title("Assigned spectra: centroid vs width")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR,"centroid_vs_width_SJZ_thicknessaware.png"), dpi=200); plt.show()
```

```python
POC_DIR        = "./sojawozniak" 
ABS_DIR        = "./sojawozniak"       
IRR_DIR        = "./sojawozniak"        
USE_SIMULATED  = True  

ABS_FILES = [
    "Cryptophytes_absortpion.csv",
    "ice_diatoms_absorption.csv",          
    "Phaeocystis_antarctica _absortpion.csv",
    "Prasinophyte_absortpion.csv"
]
IRR_FILE = ("simulated_irradiance_spectra_Antarctica.csv" 
            if USE_SIMULATED else 
            "measured_irradiance_spectra_Antarctica.csv")

# helper functions
def load_profiles():
    p_bgi = os.path.join(POC_DIR, "under_ice_bio_with_BGI.csv")
    p_v3  = os.path.join(POC_DIR, "under_ice_bio_summary_v3.csv")
    if os.path.exists(p_bgi):
        df = pd.read_csv(p_bgi, parse_dates=["time"])
    else:
        df = pd.read_csv(p_v3, parse_dates=["time"])
    if "bbp_over_chla_50_250" not in df.columns:
        with np.errstate(divide="ignore", invalid="ignore"):
            ratio = df["bbp_50_250"].values / df["chla_50_250"].values
        df["bbp_over_chla_50_250"] = ratio
        df["log10_bbp_over_chla"]  = np.log10(ratio)
    return df

def ice_bin(frac):
    z = float(frac)
    # within "under-ice" (0.15), use relative bins
    if z >= 0.95:   return "thick"
    elif z >= 0.90: return "medium"
    else:           return "thin"

def col_is_wavelength(name):
    n = str(name).strip().lower()
    return any(k in n for k in ["wavelength","lambda","nm","wl","wave"])
be
def pick_wl_column(df):
    for i, c in enumerate(df.columns):
        if col_is_wavelength(c):
            return i
    best = None
    for i in range(df.shape[1]):
        v = pd.to_numeric(df.iloc[:, i], errors="coerce")
        if v.notna().sum() > 10 and v.between(350, 750).mean() > 0.5:
            best = i; break
    return 0 if best is None else best

def read_spectra_table(path):
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam >= 380) & (lam <= 750)
    lam = lam[m]
    spectra = {}
    for j in range(df.shape[1]):
        if j == wl_idx: 
            continue
        y = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(y).sum() >= 10:
            name = str(df.columns[j]).strip()
            spectra[name] = y
    if not spectra:
        raise ValueError(f"No valid spectra columns in {os.path.basename(path)}")
    return lam, spectra

def metrics_for(lam, S):
    S = np.asarray(S, float)
    # BGI
    B = np.nanmean(S[(lam>=440)&(lam<=500)])
    G = np.nanmean(S[(lam>=520)&(lam<=560)])
    BGI = np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
    # FWHM
    Tmax = np.nanmax(S)
    if not np.isfinite(Tmax) or Tmax <= 0:
        FWHM = np.nan
    else:
        Tn = S / Tmax
        kmax = int(np.nanargmax(Tn))
        mask = Tn >= 0.5
        i0 = kmax
        while i0-1>=0 and mask[i0-1]: i0 -= 1
        i1 = kmax
        while i1+1<len(mask) and mask[i1+1]: i1 += 1
        FWHM = float(lam[i1] - lam[i0])
    # centroid (shape-weighted mean )
    Spos = np.clip(S, 0, np.inf); Ssum = np.nansum(Spos)
    centroid = float(np.nansum(lam * Spos) / Ssum) if Ssum > 0 else np.nan
    return float(BGI), float(FWHM), float(centroid)

def rank_irradiance_spectra(lam, spectra_dict):
    rows = []
    for name, S in spectra_dict.items():
        BGI, FWHM, CEN = metrics_for(lam, S)
        rows.append({"col": name, "BGI": BGI, "FWHM": FWHM, "centroid": CEN})
    df = pd.DataFrame(rows)
    # bluish: high BGI, low centroid; greenish/broad: low BGI, high centroid
    df = df.sort_values(["BGI","centroid"], ascending=[False, True]).reset_index(drop=True)
    return df

def read_absorption_file(path):
    # each file should have wavelength + absorption()
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    # find first non-wavelength numeric column
    A = None
    label = None
    for j in range(df.shape[1]):
        if j == wl_idx: continue
        v = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(v).sum() >= 10:
            A = v; label = str(df.columns[j]).strip(); break
    if A is None:
        raise ValueError(f"No numeric absorption column in {os.path.basename(path)}")
    # ensure non-negative and smooth
    A = np.clip(A, 0, np.inf)
    return lam, A, label

def resample_to(lam_src, y_src, lam_dst):
    # simple linear interpolation onto lam_dst
    y = np.interp(lam_dst, lam_src, y_src, left=np.nan, right=np.nan)
    # fill small NaN edges by nearest
    if np.isnan(y[0]):  y[0]  = y[~np.isnan(y)][0]
    if np.isnan(y[-1]): y[-1] = y[~np.isnan(y)][-1]
    return y

def compute_PUR(lam_nm, E, A):
    # integrate E() * a_ph() over  (nm); unit scaling cancels in ratios
    # : assume uniform nm spacing 
    dlam = np.median(np.diff(lam_nm))
    return float(np.nansum(E * A) * dlam)

# load profiles
tbl = load_profiles()
tbl["ice_bin"] = [ice_bin(z) for z in tbl["ice_fraction"]]

# irradiance spectra
irr_path = os.path.join(IRR_DIR, IRR_FILE)
if not os.path.exists(irr_path):
    raise FileNotFoundError(f"Irradiance file not found: {irr_path}")
lamE, Ecols = read_spectra_table(irr_path)
ranked = rank_irradiance_spectra(lamE, Ecols)
print("Irradiance spectra ranked (bluishgreenish):")
print(ranked.head(10))

# choose 3 representatives: bluest (thin), middle (medium), greenest (thick)
if len(ranked) == 1:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[0]["col"]}
elif len(ranked) == 2:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[1]["col"]}
else:
    reps = {
        "thin":   ranked.iloc[0]["col"],                   # highest BGI / bluest
        "medium": ranked.iloc[len(ranked)//2]["col"],      # middle
        "thick":  ranked.iloc[-1]["col"],                  # lowest BGI / greenest
    }
print("\nRepresentative spectra:", reps)

# assign spectrum to each profile
tbl["irr_col"] = [reps[b] for b in tbl["ice_bin"]]
# compute per-profile BGI/FWHM/centroid from assigned spectrum
BGI_list, FWHM_list, CEN_list = [], [], []
for col in tbl["irr_col"]:
    BGI, FWHM, CEN = metrics_for(lamE, Ecols[col])
    BGI_list.append(BGI); FWHM_list.append(FWHM); CEN_list.append(CEN)
tbl["BGI_ED"]         = BGI_list
tbl["FWHM_nm_ED"]     = FWHM_list
tbl["centroid_nm_ED"] = CEN_list

# absorption spectra (4 groups) 
abs_files_found = []
for f in ABS_FILES:
    # tolerate minor whitespace/typo variants in filenames
    candidates = [fn for fn in os.listdir(ABS_DIR) if fn.replace(" ","").lower()==f.replace(" ","").lower()]
    if not candidates:
        # try loose contains
        stem = re.sub(r"[^a-z]", "", f.lower())
        for fn in os.listdir(ABS_DIR):
            if stem in re.sub(r"[^a-z]","", fn.lower()):
                candidates.append(fn)
    if not candidates:
        raise FileNotFoundError(f"Absorption file not found (tried variants): {f}")
    abs_files_found.append(os.path.join(ABS_DIR, candidates[0]))

A_dict = {}   
for p in abs_files_found:
    lamA, A, label = read_absorption_file(p)
    # resample to irradiance wavelength grid
    A_res = resample_to(lamA, A, lamE)
    key = os.path.splitext(os.path.basename(p))[0]
    A_dict[key] = A_res

print("\nAbsorption spectra loaded:", list(A_dict.keys()))

# PUR per profile  phytoplankton group
for key, A in A_dict.items():
    pur_vals = []
    for col in tbl["irr_col"]:
        E = Ecols[col]
        pur_vals.append(compute_PUR(lamE, E, A))
    tbl[f"PUR_{key}"] = pur_vals

# normalize PUR by its median across profiles for readability
for key in A_dict.keys():
    med = np.nanmedian(tbl[f"PUR_{key}"].values)
    if np.isfinite(med) and med > 0:
        tbl[f"PUR_{key}_norm"] = tbl[f"PUR_{key}"] / med

# plots
from math import isnan
def safe_corr(x, y):
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() < 3: return np.nan, np.nan
    try:
        from scipy.stats import pearsonr
        r, p = pearsonr(x[m], y[m]); return float(r), float(p)
    except Exception:
        r = float(np.corrcoef(x[m], y[m])[0,1]); return r, np.nan

# BGI vs size proxy
r, p = safe_corr(tbl["BGI_ED"].values, tbl["log10_bbp_over_chla"].values)
print(f"\nBGI (assigned) vs log10(bbp/chla): r={r:.2f}, p={p if not isnan(p) else np.nan}, n={np.isfinite(tbl['log10_bbp_over_chla']).sum()}")

# PUR vs size proxy (for each phytoplankton group)
for key in A_dict.keys():
    rP, pP = safe_corr(tbl[f"PUR_{key}_norm"].values, tbl["log10_bbp_over_chla"].values)
    print(f"PUR_{key} (norm) vs log10(bbp/chla): r={rP:.2f}, p={pP if not isnan(pP) else np.nan}")

# save table
out_csv = os.path.join(POC_DIR, "poc_with_figshare_ED_irradiance_and_absorption.csv")
tbl.to_csv(out_csv, index=False)
print("\nSaved:", out_csv)

# plot: BGI vs size proxy
plt.figure()
plt.scatter(tbl["BGI_ED"], tbl["log10_bbp_over_chla"])
plt.xlabel("BGI (assigned from Antarctic irradiance spectra)")
plt.ylabel("log10(bbp/chla) [50250 m]")
plt.title("Under-ice Weddell (Oct 2020 & 2021): Light color vs size proxy")
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_BGI_vs_size.png"), dpi=200)
plt.show()

# plot: PUR (per group) vs size proxy
plt.figure()
for key in A_dict.keys():
    plt.scatter(tbl[f"PUR_{key}_norm"], tbl["log10_bbp_over_chla"], label=key)
plt.xlabel("PUR (normalized within group)")
plt.ylabel("log10(bbp/chla)")
plt.title("Under-ice: PUR vs size proxy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_PUR_vs_size.png"), dpi=200)
plt.show()

# plot: representatives spectral shape
reps_df = ranked.copy()
rep_names = [reps["thin"], reps["medium"], reps["thick"]]
plt.figure()
for nm in rep_names:
    plt.plot(lamE, Ecols[nm], label=f"{nm}")
plt.xlabel("Wavelength (nm)")
plt.ylabel("E() (arbitrary units)")
plt.title("Chosen irradiance spectra (thin=bluest, thick=greenest/broadest)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(POC_DIR, "plot_representative_spectra.png"), dpi=200)
plt.show()
```

```python
# ED35 POC

# use profiles csv detected
candidates = (glob.glob("**/under_ice_bio_with_BGI.csv", recursive=True) +
              glob.glob("**/under_ice_bio_summary_v3.csv", recursive=True))
profiles_csv = sorted(candidates, key=lambda p: (os.path.basename(p), -os.path.getmtime(p)))[0]
POC_DIR = os.path.dirname(profiles_csv) or "."
print("Using profiles CSV:", profiles_csv)

ABS_DIR = "./sojawozniak" 
IRR_DIR = "./sojawozniak"  
USE_SIMULATED = True 

# helpers
def load_profiles():
    p_bgi = os.path.join(POC_DIR, "under_ice_bio_with_BGI.csv")
    p_v3  = os.path.join(POC_DIR, "under_ice_bio_summary_v3.csv")
    df = pd.read_csv(p_bgi, parse_dates=["time"]) if os.path.exists(p_bgi) else pd.read_csv(p_v3, parse_dates=["time"])
    if "bbp_over_chla_50_250" not in df.columns:
        with np.errstate(divide="ignore", invalid="ignore"):
            ratio = df["bbp_50_250"].values / df["chla_50_250"].values
        df["bbp_over_chla_50_250"] = ratio
        df["log10_bbp_over_chla"]  = np.log10(ratio)
    return df

def ice_bin(frac):
    z = float(frac)
    if z >= 0.95:   return "thick"
    elif z >= 0.90: return "medium"
    else:           return "thin"

def pick_wl_column(df):
    for i,c in enumerate(df.columns):
        n = str(c).strip().lower()
        if any(k in n for k in ["wavelength","lambda","nm","wl","wave"]):
            return i
    # fallback guess
    for i in range(df.shape[1]):
        v = pd.to_numeric(df.iloc[:, i], errors="coerce")
        if v.notna().sum() > 10 and v.between(350,750).mean() > 0.5:
            return i
    return 0

def read_spectra_table(path):
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    spectra = {}
    for j in range(df.shape[1]):
        if j == wl_idx: continue
        y = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(y).sum() >= 10:
            spectra[str(df.columns[j]).strip()] = y
    if not spectra:
        raise ValueError(f"No valid spectra columns in {os.path.basename(path)}")
    return lam, spectra

def metrics_for(lam, S):
    S = np.asarray(S, float)
    B = np.nanmean(S[(lam>=440)&(lam<=500)])
    G = np.nanmean(S[(lam>=520)&(lam<=560)])
    BGI = np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
    Tmax = np.nanmax(S)
    if not np.isfinite(Tmax) or Tmax<=0:
        FWHM = np.nan
    else:
        Tn = S/Tmax; kmax = int(np.nanargmax(Tn))
        mask = Tn >= 0.5
        i0=kmax
        while i0-1>=0 and mask[i0-1]: i0-=1
        i1=kmax
        while i1+1<len(mask) and mask[i1+1]: i1+=1
        FWHM = float(lam[i1]-lam[i0])
    Spos = np.clip(S,0,np.inf); Ssum = np.nansum(Spos)
    centroid = float(np.nansum(lam*Spos)/Ssum) if Ssum>0 else np.nan
    return float(BGI), float(FWHM), float(centroid)

def rank_irradiance_spectra(lam, spectra_dict):
    rows = []
    for name, S in spectra_dict.items():
        BGI, FWHM, CEN = metrics_for(lam, S)
        rows.append({"col": name, "BGI": BGI, "FWHM": FWHM, "centroid": CEN})
    df = pd.DataFrame(rows).sort_values(["BGI","centroid"], ascending=[False, True]).reset_index(drop=True)
    return df

def read_absorption_file(path):
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    A, label = None, None
    for j in range(df.shape[1]):
        if j == wl_idx: continue
        v = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(v).sum() >= 10:
            A = np.clip(v, 0, np.inf); label = str(df.columns[j]).strip(); break
    if A is None:
        raise ValueError(f"No numeric absorption column in {os.path.basename(path)}")
    return lam, A, label

def resample_to(lam_src, y_src, lam_dst):
    y = np.interp(lam_dst, lam_src, y_src, left=np.nan, right=np.nan)
    if np.isnan(y[0]):  y[0]  = y[~np.isnan(y)][0]
    if np.isnan(y[-1]): y[-1] = y[~np.isnan(y)][-1]
    return y

def compute_PUR(lam_nm, E, A):
    dlam = np.median(np.diff(lam_nm))
    return float(np.nansum(E*A)*dlam)

# load profiles
tbl = load_profiles()
tbl["ice_bin"] = [ice_bin(z) for z in tbl["ice_fraction"]]

# irradiance spectra (Antarctica file)
irr_name = "simulated_irradiance_spectra_Antarctica.csv" if USE_SIMULATED else "measured_irradiance_spectra_Antarctica.csv"
irr_path = os.path.join(IRR_DIR, irr_name)
if not os.path.exists(irr_path):
    raise FileNotFoundError(f"Irradiance file not found: {irr_path}")
lamE, Ecols = read_spectra_table(irr_path)
ranked = rank_irradiance_spectra(lamE, Ecols)
print("Irradiance spectra ranked (bluishgreenish):\n", ranked.head(10))

# choose representatives: thin=bluest, medium=middle, thick=greenest
if len(ranked)==1:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[0]["col"]}
elif len(ranked)==2:
    reps = {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[1]["col"]}
else:
    reps = {"thin": ranked.iloc[0]["col"],
            "medium": ranked.iloc[len(ranked)//2]["col"],
            "thick": ranked.iloc[-1]["col"]}
print("\nRepresentative spectra:", reps)

tbl["irr_col"] = [reps[b] for b in tbl["ice_bin"]]
BGI_list, FWHM_list, CEN_list = [], [], []
for col in tbl["irr_col"]:
    BGI, FWHM, CEN = metrics_for(lamE, Ecols[col])
    BGI_list.append(BGI); FWHM_list.append(FWHM); CEN_list.append(CEN)
tbl["BGI_ED"]         = BGI_list
tbl["FWHM_nm_ED"]     = FWHM_list
tbl["centroid_nm_ED"] = CEN_list

# absorption spectra (4 groups)
abs_targets = [
    "Cryptophytes_absortpion.csv",
    "ice_diatoms_absorption.csv",
    "Phaeocystis_antarctica _absortpion.csv",
    "Prasinophyte_absortpion.csv",
]
def find_abs_file(name):
    want = re.sub(r"\s+", "", name).lower()
    for fn in os.listdir(ABS_DIR):
        if re.sub(r"\s+","",fn).lower()==want:
            return os.path.join(ABS_DIR, fn)
    # loose match by letters
    stem = re.sub(r"[^a-z]", "", want)
    for fn in os.listdir(ABS_DIR):
        if stem in re.sub(r"[^a-z]","", fn.lower()):
            return os.path.join(ABS_DIR, fn)
    raise FileNotFoundError(f"Absorption file not found: {name}")

A_dict = {}
for name in abs_targets:
    p = find_abs_file(name)
    lamA, A, label = read_absorption_file(p)
    A_res = resample_to(lamA, A, lamE)
    key = os.path.splitext(os.path.basename(p))[0]
    A_dict[key] = A_res
print("\nAbsorption spectra loaded:", list(A_dict.keys()))

# PUR per profile  group
for key, A in A_dict.items():
    tbl[f"PUR_{key}"] = [compute_PUR(lamE, Ecols[col], A) for col in tbl["irr_col"]]
for key in A_dict.keys():
    med = np.nanmedian(tbl[f"PUR_{key}"].values)
    if np.isfinite(med) and med>0:
        tbl[f"PUR_{key}_norm"] = tbl[f"PUR_{key}"]/med

# plots
def safe_corr(x,y):
    m = np.isfinite(x)&np.isfinite(y)
    if m.sum()<3: return np.nan, np.nan
    try:
        from scipy.stats import pearsonr
        r,p = pearsonr(x[m], y[m]); return float(r), float(p)
    except Exception:
        r = float(np.corrcoef(x[m], y[m])[0,1]); return r, np.nan

r,p = safe_corr(tbl["BGI_ED"].values, tbl["log10_bbp_over_chla"].values)
print(f"\nBGI (assigned) vs log10(bbp/chla): r={r:.2f}, p={p}, n={np.isfinite(tbl['log10_bbp_over_chla']).sum()}")

for key in A_dict.keys():
    rP,pP = safe_corr(tbl[f"PUR_{key}_norm"].values, tbl["log10_bbp_over_chla"].values)
    print(f"PUR_{key} (norm) vs log10(bbp/chla): r={rP:.2f}, p={pP}")

os.makedirs(POC_DIR, exist_ok=True)
out_csv = os.path.join(POC_DIR, "poc_with_ED_irr_and_abs.csv")
tbl.to_csv(out_csv, index=False); print("\nSaved:", out_csv)

plt.figure()
plt.scatter(tbl["BGI_ED"], tbl["log10_bbp_over_chla"])
plt.xlabel("BGI (Antarctic irradiance spectra)")
plt.ylabel("log10(bbp/chla) [50250 m]")
plt.title("Under-ice Weddell: Light color vs size proxy")
plt.tight_layout(); plt.savefig(os.path.join(POC_DIR,"plot_BGI_vs_size.png"), dpi=200); plt.show()

plt.figure()
for key in A_dict.keys():
    plt.scatter(tbl[f"PUR_{key}_norm"], tbl["log10_bbp_over_chla"], label=key)
plt.xlabel("PUR (normalized within group)")
plt.ylabel("log10(bbp/chla)")
plt.title("Under-ice: PUR vs size proxy")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(POC_DIR,"plot_PUR_vs_size.png"), dpi=200); plt.show()

plt.figure()
for nm in [reps["thin"], reps["medium"], reps["thick"]]:
    plt.plot(lamE, Ecols[nm], label=nm)
plt.xlabel("Wavelength (nm)"); plt.ylabel("E() (a.u.)")
plt.title("Chosen irradiance spectra (thin=bluest, thick=greenest/broadest)")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(POC_DIR,"plot_representative_spectra.png"), dpi=200); plt.show()
```

```python
# ED tables  mapping  depth window: robustness sweep

cands = (glob.glob("**/under_ice_bio_with_BGI.csv", recursive=True) +
         glob.glob("**/under_ice_bio_summary_v3.csv", recursive=True))
if not cands:
    raise FileNotFoundError("Cannot find under_ice_bio_with_BGI.csv or under_ice_bio_summary_v3.csv.")
profiles_csv = sorted(cands, key=lambda p: (os.path.basename(p), -os.path.getmtime(p)))[0]
POC_DIR = os.path.dirname(profiles_csv) or "."
print("Using profiles CSV:", profiles_csv)

ABS_DIR = "./sojawozniak"   # 4 absorption CSVs live here (we won't use PUR in this sweep)
IRR_DIR = "./sojawozniak"   # thickness-varying ED CSVs live here

# helpers
def load_profiles(size_window=(50,250)):
    """Return profiles with (re)computed size proxy for a given depth window."""
    df = pd.read_csv(profiles_csv, parse_dates=["time"])
    lo, hi = size_window
    def pick(name50_250, alt_name):
        return df[name50_250].values if alt_name not in df.columns else df[alt_name].values
    bbp = pick("bbp_50_250", f"bbp_{lo}_{hi}")
    chla= pick("chla_50_250", f"chla_{lo}_{hi}")
    with np.errstate(divide="ignore", invalid="ignore"):
        ratio = bbp / chla
        df["log10_bbp_over_chla"] = np.log10(ratio)
    # ice bin by concentration 
    def ice_bin(frac):
        z = float(frac)
        if z >= 0.95:   return "thick"
        elif z >= 0.90: return "medium"
        else:           return "thin"
    df["ice_bin"] = [ice_bin(z) for z in df["ice_fraction"]]
    return df

def pick_wl_column(df):
    for i,c in enumerate(df.columns):
        t = str(c).lower()
        if any(k in t for k in ["wavelength","lambda","nm","wl","wave"]): return i
    for i in range(df.shape[1]):
        v = pd.to_numeric(df.iloc[:, i], errors="coerce")
        if v.notna().sum()>10 and v.between(350,750).mean()>0.5: return i
    return 0

def read_irr_table(path):
    df = pd.read_csv(path)
    wl_idx = pick_wl_column(df)
    lam = pd.to_numeric(df.iloc[:, wl_idx], errors="coerce").values
    m = np.isfinite(lam) & (lam>=380) & (lam<=750)
    lam = lam[m]
    cols = {}
    for j in range(df.shape[1]):
        if j==wl_idx: continue
        y = pd.to_numeric(df.iloc[:, j], errors="coerce").values[m]
        if np.isfinite(y).sum()>=10:
            cols[str(df.columns[j]).strip()] = y
    return lam, cols

def spectrum_metrics(lam, S):
    S = np.asarray(S, float)
    B = np.nanmean(S[(lam>=440)&(lam<=500)])
    G = np.nanmean(S[(lam>=520)&(lam<=560)])
    BGI = np.nan if (not np.isfinite(B) or not np.isfinite(G) or G==0) else B/G
    Tmax = np.nanmax(S)
    if not np.isfinite(Tmax) or Tmax<=0:
        FWHM = np.nan
    else:
        Tn = S/Tmax; k = int(np.nanargmax(Tn)); mask = Tn>=0.5
        i0=k
        while i0-1>=0 and mask[i0-1]: i0-=1
        i1=k
        while i1+1<len(mask) and mask[i1+1]: i1+=1
        FWHM = float(lam[i1]-lam[i0])
    Spos = np.clip(S,0,np.inf); Ssum = np.nansum(Spos)
    centroid = float(np.nansum(lam*Spos)/Ssum) if Ssum>0 else np.nan
    return BGI, FWHM, centroid

def rank_by_color(lam, cols):
    rows=[]
    for nm,S in cols.items():
        BGI,FWHM,CEN = spectrum_metrics(lam,S)
        rows.append((nm, BGI, FWHM, CEN))
    ranked = pd.DataFrame(rows, columns=["col","BGI","FWHM","centroid"])\
               .sort_values(["BGI","centroid"], ascending=[False, True]).reset_index(drop=True)
    return ranked

def choose_representatives(ranked, strategy="extremes"):
    """Return dict {'thin':col, 'medium':col, 'thick':col} based on ranking."""
    n = len(ranked)
    if n==1:
        return {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[0]["col"], "thick": ranked.iloc[0]["col"]}
    if strategy=="extremes":
        mid = n//2
        return {"thin": ranked.iloc[0]["col"], "medium": ranked.iloc[mid]["col"], "thick": ranked.iloc[-1]["col"]}
    if strategy=="second_extremes" and n>=4:
        mid = n//2
        return {"thin": ranked.iloc[1]["col"], "medium": ranked.iloc[mid]["col"], "thick": ranked.iloc[-2]["col"]}
    if strategy=="quartiles" and n>=5:
        i10 = max(0, int(0.10*(n-1)))
        i50 = int(0.50*(n-1))
        i90 = int(0.90*(n-1))
        return {"thin": ranked.iloc[i10]["col"], "medium": ranked.iloc[i50]["col"], "thick": ranked.iloc[i90]["col"]}
    # fallback
    return choose_representatives(ranked, "extremes")

def safe_corr(x,y):
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum()<3: return np.nan, np.nan, m.sum()
    try:
        from scipy.stats import pearsonr, spearmanr
        pr, pp = pearsonr(x[m], y[m]); sr, sp = spearmanr(x[m], y[m])
    except Exception:
        pr = float(np.corrcoef(x[m], y[m])[0,1]); pp = np.nan
        sr = np.nan; sp = np.nan
    return (float(pr), float(pp), m.sum()), (float(sr), float(sp), m.sum())

# find ED files and set sweep
priority_files = [
    "scalar_irradiance_under_ice_at_euphotic_depth.csv",
    "scalar_irradiance_under_ice_at6optical_depths.csv",
    "scalar_irradiance_undericecover_at8physical_depths.csv",
]
present = []
ls = os.listdir(IRR_DIR)
for want in priority_files:
    for fn in ls:
        if fn.replace(" ","").lower()==want.replace(" ","").lower():
            present.append(fn); break
if not present:
    present = sorted([fn for fn in ls if re.sub(r"\s+","",fn.lower()).startswith("scalar_irradiance_under_ice")])

if not present:
    raise FileNotFoundError("No thickness-varying ED CSVs found in IRR_DIR.")

mapping_strategies = ["extremes", "second_extremes", "quartiles"]
size_windows       = [(50,250), (100,300)]

# sweep
rows = []
SAVE_PLOTS = False  # set True if you want per-case scatters
outdir = os.path.join(POC_DIR, "robustness_outputs"); os.makedirs(outdir, exist_ok=True)

for ed_fn in present:
    lam, cols = read_irr_table(os.path.join(IRR_DIR, ed_fn))
    ranked = rank_by_color(lam, cols)
    for strat in mapping_strategies:
        reps = choose_representatives(ranked, strat)
        for win in size_windows:
            df = load_profiles(win)
            # assign an irradiance column to each profile based on its ice_bin
            df["irr_col"] = [reps[b] for b in df["ice_bin"]]
            # compute BGI per assigned spectrum
            df["BGI_ED"] = [spectrum_metrics(lam, cols[c])[0] for c in df["irr_col"]]
            # correlations
            (pr,pp,n), (sr,sp,_) = safe_corr(df["BGI_ED"].values, df["log10_bbp_over_chla"].values)
            rows.append({
                "ed_file": ed_fn,
                "n_spectra": len(ranked),
                "mapping": strat,
                "size_window": f"{win[0]}-{win[1]}",
                "pearson_r": pr, "pearson_p": pp, "spearman_rho": sr, "spearman_p": sp, "n": n,
                "thin_repr": reps["thin"], "med_repr": reps["medium"], "thick_repr": reps["thick"]
            })
            if SAVE_PLOTS:
                plt.figure()
                plt.scatter(df["BGI_ED"], df["log10_bbp_over_chla"])
                plt.xlabel("BGI (assigned)")
                plt.ylabel("log10(bbp/chla) [size proxy]")
                plt.title(f"{ed_fn}\nmap={strat}, window={win[0]}{win[1]}  r={pr:.2f}")
                base = re.sub(r"[^a-zA-Z0-9]+","_", f"{ed_fn}_{strat}_{win[0]}_{win[1]}")
                plt.tight_layout()
                plt.savefig(os.path.join(outdir, f"BGI_vs_size_{base}.png"), dpi=160)
                plt.close()

res = pd.DataFrame(rows).sort_values(["pearson_p","spearman_p","ed_file"]).reset_index(drop=True)
res_path = os.path.join(outdir, "robustness_summary.csv")
res.to_csv(res_path, index=False)
print("\nSaved:", res_path)
print(res.head(12))
```

```python
# alternate size proxies from bbp: C:Chl and bbp/POC

POC_DIR = "poc_outputs"
os.makedirs(POC_DIR, exist_ok=True)

# 1) Find table with BGI_ED (preferred), else fall back and warn
preferred = os.path.join(POC_DIR, "poc_with_ED_thickness_varying.csv")
fallback  = os.path.join(POC_DIR, "under_ice_bio_summary_v3.csv")
if os.path.exists(preferred):
    CSV_WITH_BGI = preferred
else:
    CSV_WITH_BGI = fallback
print("Using:", CSV_WITH_BGI)

df = pd.read_csv(CSV_WITH_BGI, parse_dates=["time"], low_memory=False)

needed = ["bbp_50_250","chla_50_250"]
for c in needed:
    if c not in df.columns:
        raise ValueError(f"Column '{c}' missing. Recreate v3 table (it has bbp_50_250 & chla_50_250).")

# take BGI_ED if present (from thickness-varying ED spectra)
if "BGI_ED" in df.columns:
    bgi = df["BGI_ED"].values
    bgi_name = "BGI_ED"
else:
    # try earlier BGI column name (from first transmittance run)
    cand = [c for c in df.columns if c.upper().startswith("BGI")]
    if not cand:
        raise ValueError("No BGI column found. Run the ED thickness-varying assignment first.")
    bgi = df[cand[0]].values
    bgi_name = cand[0]

# compute POC from bbp700 using three alphas (sensitivity)
bbp = df["bbp_50_250"].values.astype(float)
chla = df["chla_50_250"].values.astype(float)

alphas = [12000.0, 13000.0, 18000.0]  # mg m^-3 per m^-1
out_rows = []

def rstats(x, y):
    from scipy.stats import pearsonr, spearmanr
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() < 3:
        return np.nan, np.nan, np.nan, np.nan, int(m.sum())
    pr, pp = pearsonr(x[m], y[m])
    sr, sp = spearmanr(x[m], y[m])
    return float(pr), float(pp), float(sr), float(sp), int(m.sum())

# store proxy arrays to plot for alpha = 13000
plots = {}

for a in alphas:
    poc = a * bbp  # mg m^-3
    with np.errstate(divide="ignore", invalid="ignore"):
        c_over_chl = poc / chla                      # C:Chl [mg C / mg Chl]
        log10_cchl = np.log10(c_over_chl)            # log10(C:Chl)
        bbp_over_poc = bbp / poc                     #  1/a 
    # correlations with BGI
    pr, pp, sr, sp, n = rstats(bgi, log10_cchl)
    out_rows.append({"proxy":"log10(POC/Chl)","alpha":a,"pearson_r":pr,"pearson_p":pp,
                     "spearman_rho":sr,"spearman_p":sp,"n":n})
    pr, pp, sr, sp, n = rstats(bgi, bbp_over_poc)
    out_rows.append({"proxy":"bbp/POC","alpha":a,"pearson_r":pr,"pearson_p":pp,
                     "spearman_rho":sr,"spearman_p":sp,"n":n})
    if a == 13000.0:
        plots["log10(POC/Chl)"] = log10_cchl
        plots["bbp/POC"] = bbp_over_poc

res = pd.DataFrame(out_rows).sort_values(["proxy","alpha"]).reset_index(drop=True)
out_csv = os.path.join(POC_DIR, "alternate_size_proxy_correlations.csv")
res.to_csv(out_csv, index=False)
print("\nSaved:", out_csv)
print(res)

# plots for alpha = 13000
def quick_scatter(x, y, xlab, ylab, title, fname):
    from scipy.stats import pearsonr
    m = np.isfinite(x) & np.isfinite(y)
    plt.figure()
    plt.scatter(x[m], y[m])
    try:
        r, p = pearsonr(x[m], y[m])
        plt.suptitle(f"r = {r:.2f}, p = {p:.3g}", y=0.98, fontsize=10)
    except Exception:
        pass
    plt.xlabel(xlab); plt.ylabel(ylab); plt.title(title)
    plt.tight_layout()
    plt.savefig(os.path.join(POC_DIR, fname), dpi=200)
    plt.show()

xlab = f"{bgi_name} (bluer  higher)"
quick_scatter(bgi, plots["log10(POC/Chl)"], xlab, "log10(POC/Chl) [50250 m]",
              "Under-ice Weddell: Color vs C:Chl (=13k)", "BGI_vs_log10POC_over_Chl.png")

quick_scatter(bgi, plots["bbp/POC"], xlab, "bbp/POC [50250 m]",
              "Under-ice Weddell: Color vs bbp/POC (=13k)", "BGI_vs_bbp_over_POC.png")
```

```python
# robustness of size proxies vs under-ice light: POC = a * bbp^b 
# csv requirements: bbp_50_250, chla_50_250, and a BGI column (BGI_ED preferred).
# optional: 'irr_col' with SojaWoniak spectrum column names (to compute FWHM/centroid).

POC_DIR = "poc_outputs"
PROFILES_CSV = None        

# locate & load the profiles table
os.makedirs(POC_DIR, exist_ok=True)
candidates = [
    PROFILES_CSV,
    os.path.join(POC_DIR, "poc_with_ED_thickness_varying.csv"),
    os.path.join(POC_DIR, "under_ice_bio_summary_v3.csv"),
    "poc_with_ED_thickness_varying.csv",
    "under_ice_bio_summary_v3.csv",
]
CSV = next((p for p in candidates if p and os.path.exists(p)), None)
if CSV is None:
    raise FileNotFoundError(
        "Could not find a profiles CSV. Set PROFILES_CSV to your file, "
        "e.g. './poc_outputs/poc_with_ED_thickness_varying.csv'"
    )

df = pd.read_csv(CSV, parse_dates=["time"], low_memory=False)
print("Using profiles file:", CSV)

for c in ["bbp_50_250", "chla_50_250"]:
    if c not in df.columns:
        raise ValueError(f"Required column '{c}' not found in {CSV}")

# choose the blue:green predictor
bgi_col = "BGI_ED" if "BGI_ED" in df.columns else None
if bgi_col is None:
    cands = [c for c in df.columns if str(c).upper().startswith("BGI")]
    if cands: bgi_col = cands[0]
if bgi_col is None:
    raise ValueError("No BGI column found (looked for 'BGI_ED'). "
                     "Run the ED thickness-varying assignment step first.")

bbp  = df["bbp_50_250"].astype(float).values
chla = df["chla_50_250"].astype(float).values
BGI  = df[bgi_col].astype(float).values
print("Predictor (light color):", bgi_col)

# (optional) compute FWHM + centroid for assigned ED spectra 
FWHM = CENT = None
irr_file = None
if "irr_col" in df.columns:
    for d in ["sojawozniak", "ED_spectra", "figshare_irradiance", "figshare", ".", POC_DIR]:
        if not os.path.isdir(d): continue
        for fn in os.listdir(d):
            nm = re.sub(r"\s+", "", fn.lower())
            if nm.startswith("scalar_irradiance_under") and nm.endswith(".csv"):
                irr_file = os.path.join(d, fn); break
        if irr_file: break

    def _pick_wl_column(df_irr):
        for i,c in enumerate(df_irr.columns):
            t = str(c).lower()
            if any(k in t for k in ["wavelength","lambda","nm","wl","wave"]):
                return i
        for i in range(df_irr.shape[1]):  # fallback heuristic
            v = pd.to_numeric(df_irr.iloc[:, i], errors="coerce")
            if v.notna().sum()>10 and v.between(350,750).mean()>0.5:
                return i
        return 0

    def _spec_metrics(lam, S):
        S = np.asarray(S, float)
        Tmax = np.nanmax(S)
        if not np.isfinite(Tmax) or Tmax<=0:
            FWHM = np.nan
        else:
            Tn = S/Tmax; k = int(np.nanargmax(Tn)); mask = Tn>=0.5
            i0=k
            while i0-1>=0 and mask[i0-1]: i0-=1
            i1=k
            while i1+1<len(mask) and mask[i1+1]: i1+=1
            FWHM = float(lam[i1]-lam[i0])
        Spos = np.clip(S,0,np.inf); Ssum = np.nansum(Spos)
        centroid = float(np.nansum(lam*Spos)/Ssum) if Ssum>0 else np.nan
        return FWHM, centroid

    if irr_file:
        try:
            irr = pd.read_csv(irr_file)
            wl_idx = _pick_wl_column(irr)
            lam = pd.to_numeric(irr.iloc[:, wl_idx], errors="coerce").values
            ok  = np.isfinite(lam) & (lam>=380) & (lam<=750)
            lam = lam[ok]
            spectra = {}
            for j in range(irr.shape[1]):
                if j==wl_idx: continue
                name = str(irr.columns[j]).strip()
                y = pd.to_numeric(irr.iloc[:, j], errors="coerce").values
                y = y[ok]
                if np.isfinite(y).sum()>=10:
                    spectra[name]=y
            fwhm, cent = [], []
            for nm in df["irr_col"].astype(str).values:
                y = spectra.get(nm)
                if y is None:
                    # try numeric match (e.g., "100.7")
                    try:
                        v = float(nm); best=None; bestd=1e9
                        for k in spectra:
                            try:
                                d = abs(float(k)-v)
                                if d<bestd: best,bestd=k,d
                            except: pass
                        y = spectra.get(best)
                    except: y=None
                if y is not None:
                    F,C = _spec_metrics(lam, y)
                else:
                    F,C = np.nan, np.nan
                fwhm.append(F); cent.append(C)
            df["FWHM_assigned"]=fwhm
            df["centroid_assigned"]=cent
            print("Computed FWHM/centroid from:", os.path.basename(irr_file))
        except Exception as e:
            print("Could not compute FWHM/centroid:", e)

# sweep a,b and compute correlations 
def corr_xy(x, y):
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum()<3: return np.nan, np.nan, np.nan, np.nan, int(m.sum())
    pr, pp = pearsonr(x[m], y[m]); sr, sp = spearmanr(x[m], y[m])
    return float(pr), float(pp), float(sr), float(sp), int(m.sum())

a_vals = np.array([8000, 10000, 12000, 13000, 15000, 18000, 20000], float)
b_vals = np.array([0.7, 0.8, 0.9, 1.0], float)

pear_heat_logCChl = np.full((len(b_vals), len(a_vals)), np.nan)
pear_heat_bbpPOC  = np.full_like(pear_heat_logCChl, np.nan)

rows=[]
for ib,b in enumerate(b_vals):
    for ia,a in enumerate(a_vals):
        POC = a * (bbp ** b)
        with np.errstate(divide="ignore", invalid="ignore"):
            logCChl = np.log10(POC/chla)
            bbpPOC  = bbp/POC
        pr1,pp1,sr1,sp1,n1 = corr_xy(BGI, logCChl)
        pr2,pp2,sr2,sp2,n2 = corr_xy(BGI, bbpPOC)
        pear_heat_logCChl[ib,ia]=pr1
        pear_heat_bbpPOC[ib,ia] =pr2
        rows.append({
            "a":a,"b":b,"n":n1,
            "pearson_r_log10POC_Chl":pr1,"pearson_p_log10POC_Chl":pp1,
            "spearman_rho_log10POC_Chl":sr1,"spearman_p_log10POC_Chl":sp1,
            "pearson_r_bbp_over_POC":pr2,"pearson_p_bbp_over_POC":pp2,
            "spearman_rho_bbp_over_POC":sr2,"spearman_p_bbp_over_POC":sp2
        })
res = pd.DataFrame(rows)
out_csv = os.path.join(POC_DIR, "grid_POCbbp_correlations.csv")
res.to_csv(out_csv, index=False)

# Heatmaps (Pearson r) 
def heatmap(data, a_vals, b_vals, title, fname):
    plt.figure()
    plt.imshow(data, aspect="auto", origin="lower")
    plt.xticks(range(len(a_vals)), [str(int(a)) for a in a_vals], rotation=45, ha="right")
    plt.yticks(range(len(b_vals)), [f"{b:.1f}" for b in b_vals])
    plt.xlabel("a in POC = a  bbp^b"); plt.ylabel("b in POC = a  bbp^b")
    plt.title(title); plt.colorbar(label="Pearson r"); plt.tight_layout()
    path = os.path.join(POC_DIR, fname); plt.savefig(path, dpi=200); plt.show()
    return path

hm1 = heatmap(pear_heat_logCChl, a_vals, b_vals,
              "r(BGI, log10(POC/Chl)) across (a,b)",
              "HM_r_BGI_vs_log10POC_over_Chl.png")
hm2 = heatmap(pear_heat_bbpPOC, a_vals, b_vals,
              "r(BGI, bbp/POC) across (a,b)",
              "HM_r_BGI_vs_bbp_over_POC.png")

# summary using FWHM/centroid as predictors (a = 13k, b = 1) 
extra_csv = None
if "FWHM_assigned" in df.columns and "centroid_assigned" in df.columns:
    a0,b0 = 13000.0,1.0
    POC0 = a0*(bbp**b0)
    with np.errstate(divide="ignore", invalid="ignore"):
        logCChl0 = np.log10(POC0/chla); bbpPOC0 = bbp/POC0
    rows2=[]
    for name,pred in [("BGI_ED", BGI),
                      ("FWHM_assigned", df["FWHM_assigned"].values),
                      ("centroid_assigned", df["centroid_assigned"].values)]:
        pr1,pp1,sr1,sp1,n1 = corr_xy(pred, logCChl0)
        pr2,pp2,sr2,sp2,n2 = corr_xy(pred, bbpPOC0)
        rows2 += [
            {"predictor":name,"proxy":"log10(POC/Chl)","pearson_r":pr1,"pearson_p":pp1,
             "spearman_rho":sr1,"spearman_p":sp1,"n":n1},
            {"predictor":name,"proxy":"bbp/POC","pearson_r":pr2,"pearson_p":pp2,
             "spearman_rho":sr2,"spearman_p":sp2,"n":n2},
        ]
    pd.DataFrame(rows2).to_csv(os.path.join(POC_DIR,"predictor_comparison_at_a13k_b1.csv"), index=False)

print("\nSaved:", out_csv)
print("Saved heatmaps:")
print(" -", hm1)
print(" -", hm2)

# display best (lowest p) combos for log10(POC/Chl)
display(res.sort_values("pearson_p_log10POC_Chl").head(8))
```
